# Seaborn library
Seaborn is a Python library used for:

* beautiful statistical graphs
* data visualization
* easier plotting than matplotlib
* analyzing datasets visually

It is built on top of Matplotlib.

### Why Seaborn?
* provides a layer of abstraction hence simpler to use -- less lines of codes with efficient results
* better aesthetics
* more graphs included
* famous

### Seaborn Roadmap
#### Types of Function/Graphs
* Figure Level
* Axis Level

(like two objects in matplotlib)

#### Main Classification
* Relational Plot --- Scatter plot, line plot
* Distribution Plot --- Histogram, kdeplot, Rugplot
* Categorical Plot --- bar plot, count plot, box plot, violin, swamplot 
* Regression Plot --- (linear regression)
* Matrix Plot --- Heatmap, clustermap
* Multiplots --- Joint plot, Pair plot

Documentation of seaborn:
https://seaborn.pydata.org/api.html





## 1. Relational Plot
--- scatter plot, line plot
* to see the statistical relation between 2 or more variables.
* Bivariate Analysis (2 variables, 2 columns)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

### scatter plot

```python
# scatter plot -> axis level function
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

### line plot 
--- to show continuity eg date, time

```python
gap = px.data.gapminder()
gap
```

```python
temp_df = gap[gap['country'] == 'India']
temp_df 
```

```python
# axis level function

sns.lineplot(data = temp_df, x = 'year', y = 'lifeExp')
```

```python
# figure level function

sns.relplot(data= temp_df, x='year', y='lifeExp', kind = 'line')
```

```python
sns.relplot(data= temp_df, x='year', y='lifeExp', kind = 'line', hue = 'country')
```

```python
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country', style='continent', size='continent')
     
```

```python
# facet plot
sns.relplot(data = tips, x='total_bill', y = 'tip', kind = 'scatter', col = 'sex')
```

```python
# facet plot
sns.relplot(data = tips, x='total_bill', y = 'tip', kind = 'scatter', row = 'sex')
```

```python
# facet plot --- work only with figure level function
# it will not work with scatterplot and lineplot
sns.relplot(data = tips, x='total_bill', y = 'tip', kind = 'scatter', row = 'sex', col='smoker')
```

col/row is used when we want to have individual graph of a particular column 

```python
# col wrap
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=3)
     
```

##  2. Distribution Plot
-- tells how data is distributed
* used for univariate analysis
* used to find out the distribution
* Range of the observation
* Central Tendency
* is the data bimodal? -- (multiple peaks)
* Are there outliers?

Plots under distribution plot

* histplot
* kdeplot
* rugplot

##### Levels:
* figure level ---> displot
* axis level ---> histplot, kdeplot, rugplot

### histplot

```python
# plotting univariate histogram
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind = 'hist')
```

```python
sns.displot(data=tips, x='total_bill', kind = 'hist', bins = 20)
```

It’s also possible to visualize the distribution of a categorical variable using the logic of a histogram.

Discrete bins are automatically set for categorical variables


```python
# countplot
sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist',hue='sex')
```

```python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist',hue='sex', element='step')
```

```python

titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x='age', kind='hist',element='step',hue='sex')
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(data=tips, x='tip', kind='hist',col='sex',element='step')
     
```

### kdeplot

Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate

way of prediction --- can tell between values as well

```python
# axis level
sns.kdeplot(data=tips,x='total_bill')
```

```python
# figure level
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue -> fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex')
     
```

```python
# hue -> fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=10,aspect=2)
     
```

### Rug plot

Plot marginal distributions by drawing ticks along the x and y axis

This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

     

```python
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

#### Bivariate histogram
 A bivariate histogram bins the data within rectangles that tile the plot and then shows the count of observations within each rectangle with the fill color



```python
sns.histplot(data=tips, x='total_bill', y='tip')
```

```python
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
```

#### Bivariate Kdeplot
a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian

```python
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

## 3. Matrix Plot

* Heatmap
* Clustermap

### Heatmap
 Plot rectangular data as a color-encoded matrix

```python
gap
```

```python
gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')
```

```python

# axes level function
sns.heatmap(temp_df)
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# annot
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5, cmap='summer') #annot for numbers(value amount) , gap by linewidth
```

### Clustermap

-- clustering -- similar data together.
Plot a matrix dataset as a hierarchically-clustered heatmap.

(This function requires scipy to be available.)

```python
iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```

Organized columns according to clustering
