# Group-by

groupby() in pandas is one of the most important concepts for data analysis. It’s used to split data into groups, apply operations, and combine results.

👉 groupby() means:

> “Group similar data together, then perform calculations on each group separately.”

### Categorical VS Numerical data

Categorical data = data that represents labels / groups / categories (not numbers you calculate with).

It answers:

“Which type / group does this belong to?”

eg, gender, age, city, grade, section, subscription, payment method

and Numerical says "How much?"

1. Nominal (NO ORDER)

👉 Just labels

* gender
* city
* country

❌ No ranking

2. Ordinal (HAS ORDER)

👉 Order matters

* grades: A > B > C
* ratings: 1 ⭐ to 5 ⭐
* education: school → college → university

👉 Still NOT numeric, but ordered


| Column Type | How to fill      |
| ----------- | ---------------- |
| Numeric     | mean / median    |
| Categorical | mode / "Unknown" |


#### DataFrame

```python
import pandas as pd

data = {
    "name": ["Ali", "Sara", "Noor", "Zain", "Hina"],
    "city": ["Lahore", "Lahore", "Karachi", "Karachi", "Lahore"],
    "marks": [80, 90, 85, 70, 95],
    "age": [20, 21, 22, 23, 21]
}

df = pd.DataFrame(data)
df
```

#### Group-by one column

* Average marks per city



```python
df.groupby("city")["marks"].mean()

```

#### Group by multiple columns

```python
df.groupby(["city", "age"])["marks"].mean()
```

 
### Multiple Aggregate Functions

* Using Dict:

> df.groupby().agg({"col name": "func" , "col name2": "func"})

specific Fnction on specfic col

* using list:

> df.groupby().agg( ["func", "func2] )

apply these function on ever col

* merge

> df.groupby.agg( {"col name": ["func1" , "func2" ]} )

```python
df.groupby("city")["marks"].agg(["mean", "sum", "count"])
```

#### Grouping multiple columns

```python
df.groupby("city")[["marks", "age"]].mean()
```

#### Size of each group
tells how many rows in each group

```python
df.groupby("city").size()
```

#### Get group as DataFrame

```python
df.groupby("city").get_group("Lahore")
#returns only Lahore data
```

#### Iterating groups

```python
for name, group in df.groupby("city"):
    print(name)
    print(group)
# prints each group separately
```

### Using custom aggregation

```python
df.groupby("city").agg({
    "marks": "mean",
    "age": "max"
})
```

> Key concept: Split → Apply → Combine

groupby() works in 3 steps:

1. Split

Divide data into groups

2. Apply

Perform function (mean, sum, etc.)

3. Combine

Return final result

```python
import pandas as pd
df = pd.read_csv("dirty_cafe_sales.csv")
df.head()
```

```python
print(df['Price Per Unit'].dtype)
```

```python
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce') 
```

```python
print(df['Price Per Unit'].dtype)
```

```python
grouped_mean= df.groupby('Item')['Price Per Unit'].mean()
print(grouped_mean)
```

Group the data by Item, then for each item, calculate the average (mean) price per unit.

```python
grouped_sum = df.groupby(['Item' ,'Location'])['Price Per Unit'].sum()
print(grouped_sum)
```

“Group my data by Item AND Location, then for each group, add up (sum) the Price Per Unit.”

	
So instead of treating every row separately, it clusters rows into mini-groups.

After grouping, you say:

👉 “From each group, I only care about this column”


#### 📊 Example 

##### Original Data:

| Item   | Location | Price Per Unit |
| ------ | -------- | -------------- |
| Coffee | In       | 2              |
| Coffee | In       | 3              |
| Coffee | Take     | 2              |
| Cake   | In       | 3              |

---

##### After grouping + sum:

| Item   | Location | Sum |
| ------ | -------- | --- |
| Coffee | In       | 5   |
| Coffee | Take     | 2   |
| Cake   | In       | 3   |

---




```python
grouped_sum = df.groupby(['Item', 'Location'])['Price Per Unit'].sum().reset_index()
print(grouped_sum)
```

```python
grouped_agg = df.groupby('Location')['Price Per Unit'].agg(['mean', 'sum', 'count'])
grouped_agg
```

👉 “Group data by Location, and for each location, calculate:

* average price
* total price
* number of entries”

.agg means instead of just one operation, you **apply multiple operations** at once:

| Function | Meaning                   |
| -------- | ------------------------- |
| mean     | average                   |
| sum      | total                     |
| count    | number of non-null values |

### Multiple Aggregate Functions

* Using Dict:

> df.groupby().agg({"col name": "func" , "col name2": "func"})

specific Fnction on specfic col

* using list:

> df.groupby().agg( ["func", "func2] )

apply these function on ever col

* merge

> df.groupby.agg( {"col name": ["func1" , "func2" ]} )

```python
genre.sum().head(1) 
```

```python
# apply specific function on specifc col

genre.agg( 
    {
        "Runtime": "mean",
        "IMDB_Rating": "mean",
        "No_of_Votes": "sum",
        "Gross": "sum",
        "Metascore": "min"
    }
)
```

```python
genre.agg( ["max", "min"])
# now each col has min max col
```

```python
genre.agg( 
    {
        "Runtime": ["min", "max"],
        "IMDB_Rating": ["min", "max"],
        "No_of_Votes": ["sum", "mean"],
        "Gross" :    ["sum", "max", "min"]
    }
)
```

```python
# find Nmbr of movies starting with A for each group 

def starts_A(group):
    return group["Series_Title"].str.startswith('A').sum()

genre.apply(starts_A)
```

```python
def normal(group):
    group["norm_rating"] = group["IMDB_Rating"].min() / (group["IMDB_Rating"].max() - group["IMDB_Rating"].min())
    return group

genre.apply(normal)
```

```python
# which actor genre combo is best in meta score

movies.groupby(["Genre", "Star1"])["Metascore"].mean()
```
