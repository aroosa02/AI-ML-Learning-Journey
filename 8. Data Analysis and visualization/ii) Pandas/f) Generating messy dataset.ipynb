# Generate messy dataset

This is a real-world data generation script used to create a messy dataset for practice.

```python
import numpy as np
import pandas as pd
import random
from faker import Faker 

fake = Faker()
np.random.seed(42)
random.seed(42)

n_rows = 1000

# Generate messy data
data = {
    "user_id": [fake.uuid4() for _ in range(n_rows)],
    "age": np.random.randint(18, 70, n_rows),
    "gender": np.random.choice(["Male", "Female", "male", "female", "M", "F", None], n_rows),
    "country": np.random.choice(["USA", "UK", "India", "Pakistan", None, "us", "UK"], n_rows),
    "signup_date": [fake.date_between(start_date="-5y", end_date="today") for _ in range(n_rows)],
    "last_login": [fake.date_between(start_date="-2y", end_date="today") for _ in range(n_rows)],
    "subscription_plan": np.random.choice(["Free", "Basic", "Premium", None], n_rows),
    "purchase_amount": np.random.choice([np.nan, *np.random.randint(0, 500, 10)], n_rows),
    "feedback_score": np.random.choice([1, 2, 3, 4, 5, None], n_rows),
    "referrer": np.random.choice([fake.name(), None, "Friend", "Google", "Ad", "ad"], n_rows),
    "notes": [fake.sentence(nb_words=6) if random.random() > 0.2 else None for _ in range(n_rows)],
    "irrelevant_col": np.random.rand(n_rows),
}

df = pd.DataFrame(data)

# Introduce some duplicates
df = pd.concat([df, df.sample(50)], ignore_index=True)

# Shuffle rows
df = df.sample(frac=1).reset_index(drop=True)

# Save the dataset for practice
df.to_csv("messy_dataset.csv", index=False)

print("Messy dataset generated with shape:", df.shape)
print(df.head(10))
```

2. seed()

The purpose of seed is to make randomness repeatable (not actually random each time).

> random.seed(42)

you are telling Python:

“Start generating random numbers from a fixed starting point.”

Normally, random values change every time you run code:

Run 1 → different numbers

Run 2 → different numbers

Run 3 → again different

But with a seed:

👉 You get the same “random” output every single time

```python
import pandas as pd
import numpy as np

np.random.seed(42)

characters = [
    "Mickey","Donald","Goofy","SpongeBob","Patrick","Bugs Bunny",
    "Daffy Duck","Tom","Jerry","Scooby-Doo","Shaggy",
    "Fred","Velma","Daphne","Popeye","Olive","Blossom","Bubbles","Buttercup","Ben10"
]

studios = ["Disney","Nickelodeon","Warner Bros","Cartoon Network","Universal"]

teams = ["Mystery Inc","Looney Tunes","Powerpuff Girls","Avengers Kids","Toon Squad"]

roles = ["Sidekick","Hero","Villain","Leader","Comic Relief"]

df = pd.DataFrame({
    "Character": np.random.choice(characters,500),
    "Studio": np.random.choice(studios,500),
    "Team": np.random.choice(teams,500),
    "Role": np.random.choice(roles,500),
    "PowerLevel": np.random.randint(50,1000,500),
    "Wins": np.random.randint(0,100,500),
    "Losses": np.random.randint(0,50,500),
    "Skills": np.random.randint(5,200,500),
    "Age": np.random.randint(5,60,500)
})

df.head(10)
```
