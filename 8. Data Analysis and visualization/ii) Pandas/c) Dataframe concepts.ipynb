# DataFrame 

A dataframe = table (rows + columns)

It is most important pandas structure

### 1) Creating Dataframe

```python
import pandas as pd
import numpy as np

# From dictionary
df = pd.DataFrame({
    "Name": ["Aroosa", "Noor", "Qurrat"],
    "Age": [21, 22, 23]
})
df



```

```python
# From list of lists
df2 = pd.DataFrame([
    ["Aroosa", 21],
    ["Noor", 22],
    ["Qurrat", 23]
], columns=["Name", "Age"])
df2
```

```python
# From NumPy array
arr = np.array([[1, 2], [3, 4]])
df3 = pd.DataFrame(arr, columns=["A", "B"])
df3
```

```python

# Empty DataFrame
df4 = pd.DataFrame()
df4
```

### 2) Basic Attributes

```python
print(df.shape)      # (rows, columns)
print(df.size)       # total elements
print(df.ndim)       # dimensions (2)
print(df.columns)    # column names
print(df.index)      # row labels
print(df.dtypes)     # data types
print(df.values)     # numpy array
```

```python
df.shape[0] # number of rows by default
```

### 3) Viewing data

```python
print(df.head())        # first 5 rows
print(df.tail())        # last 5 rows
print(df.sample(2))     # random rows
print(df.info())        # summary
print(df.describe())    # statistics
```

### 4) Selecting Columns

```python
print(df["Name"])         # single column
df[["Name", "Age"]]            # multiple columns
```

### 5) Selecting rows (imp)

```python
# By position
print(df.iloc[0]  )      # first row
print(df.iloc[0:2]  )    # first two rows

# By label
print(df.loc[0]  )       # first row
print(df.loc[0:2]  )     # first two rows

# Specific row & column
print(df.iloc[0, 1])
print(df.loc[0, "Age"])
```

### 6) Filtering Data

```python
print(df[df["Age"] > 21])   # filter rows
print()
print(df[df["Name"].str.startswith("A")])  # string filter
print()
print(df[df["Name"].str.endswith("a")])    # string filter
print()
print(df[df["Name"].str.contains("o")])  # string contains
print()
print(df[df["Name"].isin(["Aroosa", "Qurrat"])])  # multiple values
print()
# Multiple conditions
print(df[(df["Age"] > 20) & (df["Age"] < 23)])
```

### 12) Handling Missing Values

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({ #include null values
    "Name": ["Aroosa", "Noor", np.nan, "Qurrat"],
    "Age": [25, np.nan, 30, 35],
    "Salary": [50000, 60000, 70000, np.nan]
})
df
```

```python
df.isnull()
```

```python
df.notnull()
```

```python
df.dropna()          # remove missing rows
```

```python
df.fillna(0)         # replace with value
```

```python
df.fillna(df.mean()) # replace with mean but can only concatenate numeric columns
```

### 20) GroupBy

```python
df = pd.DataFrame({
    "Dept": ["IT", "IT", "HR", "HR"],
    "Salary": [1000, 2000, 1500, 2500]
})
df
```

```python
df.groupby("Dept").sum() # group by and sum
```

```python
 df.groupby("Dept").mean() # group by and calculate mean
```

### 21) Merge (like SQL JOIN)

```python
df1 = pd.DataFrame({"ID": [1,2], "Name": ["Aroosa","Kashaf"]})
df1
```

```python
df2 = pd.DataFrame({"ID": [1,2], "Salary": [1000,2000]})
df2
```

```python
pd.merge(df1, df2, on="ID") # merge on common column
```

### 22) Concatenate

```python
pd.concat([df1, df2])
```

### 23) Pivot Table
 pivot table is used to summarize data by grouping and applying an aggregation function. In this example, we are calculating the mean salary for each department.

```python
df.pivot_table(values="Salary", index="Dept", aggfunc="mean")

```
