# Series

series is 1d array structure,
column or row

that can hold any datatype.

```python
import pandas as pd

data = [10, 20, 30, 40]
s = pd.Series(data)

print(s)
print(type(s))
```

```python
# the other way to create a Series is to use a dictionary

data = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
s_dict = pd.Series(data)
print(s_dict)
print(type(s_dict))
# indexes are replaced by dict keys
```

```python
# without using dictionary, we can also specify the index of the Series
data = [10, 20, 30, 40]
index = ['a', 'b', 'c', 'd']
s_index = pd.Series(data, index=index)
print(s_index)
```

### 1) Creating a series

```python
#so we can create a series in multiple ways

# From list
s1 = pd.Series([10, 20, 30])
print(s1)

# With index
s2 = pd.Series([10, 20, 30], index=["a", "b", "c"])
print(s2)

# From dictionary
s3 = pd.Series({"a": 1, "b": 2, "c": 3})
print(s3)

# From scalar (same value repeated)
s4 = pd.Series(5, index=[0, 1, 2])
print(s4)
```

### 2) Basic attributes of series

```python
s = pd.Series([10, 20, 30, 40])

print(s.values)      # array([10, 20, 30, 40])
print(s.index)       # RangeIndex(start=0, stop=4, step=1)
print(s.dtype)       # data type
print(s.shape)       # (4,)
print(s.size)        # 4
print(s.ndim)        # 1
print(s.name)        # name of series
```

### 3) Accessing data

```python
s = pd.Series([10, 20, 30], index = ['a', 'b', 'c'])

print(s['a']) # to access a value using index label
print(s[0]) # to access a value using index position

print(s[['a','c']]) # to access multiple values using index labels
print(s[[0,2]]) # to access multiple values using index positions

print(s[0:2]) # to access a range of values using index positions (exclusive of end index)
print(s['a':'c']) # to access a range of values using index labels (inclusive of end index)
```

### 4) loc and iloc in series

```python
s = pd.Series([10, 20, 30, 40], index = ['a', 'b', 'c', 'd'])
print(s)

s.loc['a'] # access using label (label-based)
print(s)
s.iloc[0] # access using position (position-based)
print(s)
```

### 5) Modifying data

```python
# Update values in a Series

s["a"] = 100 # update value using label
print(s)

s[0] = 200 # update value using position
print(s)
```

```python
# Add values to a Series

s["e"] = 50 # add value using label
print(s)

s[4] = 60 # add value using position
print(s)
```

### 6) Mathematical operations

```python
s = pd.Series([1, 2, 3, 4, 5])
print(s)

print(s + 5)
print(s * 2)
print(s ** 2)
print(s / 2)
print(s - 1)

# works element-wise just like numpy arrays
```

### 7) Comparison operators

```python
s = pd.Series([10, 20, 30, 40])
print(s)

print(s > 20) # returns boolean series
print(s[s > 20]) # filter values greater than 20
```

### 8) Aggregation functions

```python
s = pd.Series([10, 20, 30, 40])

print(f"Sum: {s.sum()}") # sum of all values
print(f"Mean: {s.mean()}") # mean of values
print(f"Minimum: {s.min()}") # minimum value
print(f"Maximum: {s.max()}") # maximum value
print(f"Count: {s.count()}") # count of non-null values
print(f"Median: {s.median()}") # median value
print(f"Standard Deviation: {s.std()}") # standard deviation
print(f"Variance: {s.var()}") # variance
```

### 9) Value Based functions

```python
s = pd.Series([1, 2, 9, 3, 0, 4, 5, 2, 7, 8, 0, 1, 2 ])
print(s)

print(f"Value Counts:\n {s.value_counts()}") # count of unique values
print(f"Relative Frequencies:\n {s.value_counts(normalize=True)}") # relative frequencies of unique values
print(f"Sorted Value Counts:\n {s.value_counts().sort_index()}") # count of unique values sorted by index
print(f"Sorted Value Counts (descending):\n {s.value_counts().sort_values(ascending=False)}") # count of unique values sorted by count (descending)

print(f"Unique Values:\n {s.unique()}") # unique values in the series
print(f"Number of Unique Values:\n {s.nunique()}") # number of unique values in the series
```

###  10) Sorting

```python
print(s.sort_values()) # sort values in ascending order
print()
print(s.sort_values(ascending=False)) # sort values in descending order
print()
# print(s) # original series remains unchanged

print(s.sort_index()) # sort by index
print()
print(s.sort_index(ascending=False)) # sort by index in descending order
```

### 11) Handling missing values

```python
import numpy as np

s = pd.Series([1, 2, np.nan, 4, 5, np.nan])
print(s)
print()

print(s.isnull()) # check for null values
print()
print(s.notnull()) # check for non-null values
print()

print(s.dropna()) # drop null values
print()
print(s.fillna(0)) # fill null values with 0
print()
print(s.fillna(method='ffill')) # forward fill null values (fill with previous value)
print()
print(s.fillna(method='bfill')) # backward fill null values (fill with next value)
```

### 12) Index Operations

```python
s = pd.Series([10, 20, 30], index = ['a', 'b', 'c'])
print(s)
print()

# Change index
s.index = ['x', 'y', 'z']
print(s)
print()

# Reset index
s.reset_index(drop=True, inplace=True) # drop=True to remove old index, inplace=True to modify original series
print(s)
print()

# Check index
print(s.index)
```

### 21) Index Alignment Concept
when performing operations on two series, pandas aligns them based on their index labels. If an index label is present in one series but not the other, the result will be NaN for that label.

```python
s1 = pd.Series([1, 2], index=["a", "b"])
s2 = pd.Series([3, 4], index=["b", "c"])

s1 + s2
```
