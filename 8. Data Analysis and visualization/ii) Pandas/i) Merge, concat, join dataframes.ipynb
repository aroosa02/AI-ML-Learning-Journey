# Concatenation, Join & Merge


| Method     | Think of it as                             |
| ---------- | ------------------------------------------ |
| `concat()` | stacking data (top/bottom or side-by-side) |
| `merge()`  | SQL-style join using keys                  |
| `join()`   | shortcut for merge (index-based)           |


### **1. concat() → “Stack or glue DataFrames”**

-> Use when:
* DataFrames have same columns
* You want to append rows or combine columns

```python
import pandas as pd

# row-wise concatenation
df1 = pd.DataFrame({"A": [1, 2], "B": [3, 4]})
df2 = pd.DataFrame({"A": [5, 6], "B": [7, 8]})

pd.concat([df1, df2])
```

```python
# column-wise concatenation
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1, df2], keys = ["First", "Second"])
```

concat = just stick DataFrames together (no matching logic)

### **2. merge() → “Database-style join”**

-> Use when:
* You have common column (key)
* Like SQL JOIN

```python
df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Aroosa", "Noor", "Qurrat"]
})

df2 = pd.DataFrame({
    "id": [0, 2, 3],
    "marks": [80, 90, 70]
})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on="id")
```

👉 “Combine df1 and df2 using the common column 'id', and keep only matching rows.”


* on="id"

Think of it like:

“Join rows where df1.id == df2.id”

#### **Types of join**

| Type  | Meaning            |
| ----- | ------------------ |
| inner | only matching rows |
| left  | all rows from df1  |
| right | all rows from df2  |
| outer | all rows from both |

##### 1. Inner (default) ---intersection

```python
pd.merge(df1, df2, on="id", how="inner")
```

* how="inner"
 
👉 This decides what type of merge

INNER JOIN means:

Only keep rows where the key exists in BOTH dataframes

##### 2. Left join

```python
pd.merge(df1, df2, on="id", how="left")
```

Analogy: left → “I trust df1, fill from df2 if possible”


##### 3. Right join

```python
pd.merge(df1, df2, on="id", how="right")
```

Analogy: right → “I trust df2, fill from df1 if possible”

##### 4. Outer join --- union

```python
pd.merge(df1, df2, on="id", how="outer")
```

Analogy: outer → “give me EVERYTHING”

### **3. join() → “Index-based merge (shortcut)”**

-> Use when:
* joining on index
* simpler syntax

```python
df1 = pd.DataFrame({"A": [1, 2]}, index=["a", "b"])
df2 = pd.DataFrame({"B": [3, 4]}, index=["a", "b"])

df1.join(df2)
```

```python
df1.join(df2, how="inner")
```

👉 If you need matching → use merge

👉 If you just want to stack → use concat
