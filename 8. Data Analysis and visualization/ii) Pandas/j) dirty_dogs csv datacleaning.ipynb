```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dogs_dirty_dataset.csv')
df
```

```python
print(df.shape)
```

```python
print(df.info())
```

```python
# Check missing values
print(df.isnull().sum())
```

### Step 1
Right now this dataset has:

"missing"

"error"

"unknown"

actual NaN

```python
# Replace "missing", "error", and "unknown" with NaN
df.replace(["missing", "error", "unknown"], pd.NA, inplace=True)
```

### Step 2
Columns like:

weight_kg

height_cm

owner_age

daily_walk_minutes

barks_per_day

are object but should be numeric

```python
cols_to_numeric = [
    "weight_kg", "height_cm", "owner_age",
    "daily_walk_minutes", "barks_per_day",
    "vet_visits_per_year"
]

for col in cols_to_numeric:
    df[col] = pd.to_numeric(df[col], errors='coerce') # errors = coerce will convert invalid parsing to NaN
```

```python
df.info()
```

### Step 4
 #### Fix Invalid Values

Example:

age_years = -1 -- impossible

```python
df.loc[df["age_years"] < 0, "age_years"] = pd.NA
```

```python
df.describe()
```

### Step 5
#### Covert date column

Columns:

last_vet_visit

adoption_date

```python
df["last_vet_visit"] = pd.to_datetime(df["last_vet_visit"], errors='coerce')
df["adoption_date"] = pd.to_datetime(df["adoption_date"], errors='coerce') # errors = coerce will convert invalid dates to NaT (Not a Time)
```

### Step 6
#### Clean categorical data
Example: vaccinated, neutered

```python
df["vaccinated"].value_counts()
```

```python
df["vaccinated"] = df["vaccinated"].str.lower().str.strip()
df["neutered"] = df["neutered"].str.lower().str.strip()
```

```python
df["vaccinated"] = df["vaccinated"].map({"yes": 1, "no": 0})
df["neutered"] = df["neutered"].map({"yes": 1, "no": 0})
```

```python
df.isnull().sum()
```

### Step 7
####  Handle missing values
Option A: Drop rows (if too many missing)

Option B: Fill numeric values

Option C: Fill categorical

```python
# option A: drop rows with missing dog names
df.dropna(subset=["dog_name"], inplace=True)
```

```python
df.isnull().sum()
```

```python
# option B: 
df["age_years"] =df["age_years"].fillna(df["age_years"].median(), inplace=True)
df["weight_kg"] = df["weight_kg"].fillna(df["weight_kg"].mean(), inplace=True)
```

```python
# option C:
df["breed"].fillna("Unknown", inplace=True)
```

### Step 8:
#### Fix Sensor Columns
 

sensor_metric_1 → float 

sensor_metric_21+ → object  (dirty)

Convert them:

```python
sensor_cols = [col for col in df.columns if "sensor_metric_" in col]

for col in sensor_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

### Step 9
#### Remove duplicates


```python
df.drop_duplicates(inplace=True)
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

### Decide drop or keep

Rule:
* If >50% missing → consider dropping column
* If important → fill smartly

```python
df.drop(columns=["age_years", "weight_kg"], inplace=True)
```

```python
# df["height_cm"].fillna(df["height_cm"].median(), inplace=True)
df["height_cm"] = df["height_cm"].fillna(df["height_cm"].median())
```

```python
df["color"]= df["color"].fillna(df["color"].mode()[0])
```

```python
df["vaccinated"] = df["vaccinated"].fillna(df["vaccinated"].mode()[0])
```

```python
df["neutered"] = df["neutered"].fillna(df["neutered"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["last_vet_visit", "adoption_date"], inplace=True)
# or
# df["last_vet_visit"].fillna(method="ffill", inplace=True)
```

### Remove unrealistic values

```python
df.loc[df["owner_age"] < 10, "owner_age"] = pd.NA
df.loc[df["owner_age"] > 100, "owner_age"] = pd.NA
```
