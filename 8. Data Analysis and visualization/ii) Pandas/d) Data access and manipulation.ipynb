#### loc() --- access row index

```python
df.loc[0]  # to access a row of the dataframe use loc[row index] method
```

#### iloc() --- index location

```python
df.iloc[0]  # to access a row of the dataframe use iloc[row index] method   
```

```python
df.loc[0][0]
```

difference between loc and iloc is 

* .loc is label based index

eg df.loc[:, 'A']  --- column named A

* It accepts conditions.
* Includes end index when slicing.

* .iloc is position based index

eg df.iloc[:, 0]  --- first column

* It does not accepts conditions.
* Excluded end index.

Use .loc when you know the labels (names)

Use .iloc when you know the positions (numbers)



```python
df.iloc[0][2]
```

### Data Manipulation with Dataframe

##### Add a new column to the dataframe

```python

df['Salary'] = [50000, 60000, 70000]
df
```

##### Remove a column from the dataframe


```python
# Remove a column from the dataframe

#df.drop('Salary') # error since it will check in row first axis = 0 which does not a salary
df.drop('Salary', axis=1) # to remove a column we have to specify axis = 1
```

drop is a temporary operation not a permanent operation it does not save state of dataframe so when running next time it will appear again

```python
df
```

In order to make it save it we use inplace = True

```python
df.drop('Salary', axis=1, inplace=True) # to remove a column permanently we have to specify inplace = True
```
