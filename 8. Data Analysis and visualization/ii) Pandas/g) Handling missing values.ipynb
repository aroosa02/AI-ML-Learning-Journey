# Handling missing values

None and np.nan (both treated as missing)

```python
import pandas as pd
import numpy as np

data = {
    "name": ["Aroosa", "Ayesha", None, "Noor", None, "Maheen"],
    "age": [20, np.nan, 22, 21, np.nan, 23],
    "marks": [85, 90, np.nan, 88, 76, np.nan],
    "city": ["Islamabad", None, "Lahore", "Karachi", None, "Islamabad"]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum() # Count how many missing values are in each column
```

👉 This tells you:

which columns have missing values

how many in each

```python
df.dropna()
```

```python
df.dropna(subset=["marks"])
```

dropna(how="all") means:

* It removes only those rows where all values are missing (NaN)

* If even one value exists, the row stays.

1. Remove rows with any missing values

> df.dropna()

2. Remove rows where specific column is missing

> df.dropna(subset=["name"])

3. Remove columns with all NaN

> df.dropna(axis=1, how="all")

```python
df.fillna(0)
```

```python
df["age_nonNull"] =df["age"].fillna(df["age"].mean())
df
```

```python
df. drop('age_nonNull',axis=1, inplace=True)
```

```python
# Forward fill
df.fillna(method="ffill")
```

```python
# Backward fill
df.fillna(method="bfill")
```

```python
# percentage of missig Value in eac col
(df.isnull().sum() / len(df)) * 100
```

```python
# Visualize missing Values
df.isnull().sum().sort_values(ascending = False) #descending order
```

```python
# fill missing Vales
df["filled_age"] = df["age"].fillna(21)
df
```

```python
df["city"].sample(3) # Randomly sample 3 values from the "city" column
```

```python
# Drop columns if >40% missing

threshold = len(df) * 0.5

df = df.dropna(axis = 1, thresh = threshold) # Drop columns with more than 50% missing values
```

```python
# Fill Missing Perfromance using mean

df["performance_score"] = df["marks"].fillna(df["marks"].mean())
df
```

```python
df.isna().sum() 
```

### Now we'll handle all missing values from the dataframe

### name column

```python
df["name"] = df["name"].fillna("Unknown")
df
```

### age column

```python
df["age"] = df["age"].fillna(df["age"].mean())
df
```

### marks column

```python
df["marks"] = df["marks"].fillna(df["marks"].mean())
df
```

### city column

#### Option A (most common in exams):

```python
df["city"] = df["city"].fillna("Unknown")
df
```

#### Option B (better in real life):

```python
df["city"] = df["city"].fillna(df["city"].mode()[0])
df
```

mode() = most frequent value
