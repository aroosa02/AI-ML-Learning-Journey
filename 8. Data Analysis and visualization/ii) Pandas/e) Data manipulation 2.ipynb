### Data Aggregating and Grouping

```python
df.head()
```

```python
grouped_mean= df.groupby('Item')['Price Per Unit'].mean()
print(grouped_mean)
```

“Group the data by Item, then for each item, calculate the average (mean) price per unit.”

```python
grouped_sum = df.groupby(['Item' ,'Location'])['Price Per Unit'].sum()
print(grouped_sum)
```

“Group my data by Item AND Location, then for each group, add up (sum) the Price Per Unit.”

	
So instead of treating every row separately, it clusters rows into mini-groups.

After grouping, you say:

👉 “From each group, I only care about this column”


#### 📊 Example 

##### Original Data:

| Item   | Location | Price Per Unit |
| ------ | -------- | -------------- |
| Coffee | In       | 2              |
| Coffee | In       | 3              |
| Coffee | Take     | 2              |
| Cake   | In       | 3              |

---

##### After grouping + sum:

| Item   | Location | Sum |
| ------ | -------- | --- |
| Coffee | In       | 5   |
| Coffee | Take     | 2   |
| Cake   | In       | 3   |

---




```python
grouped_sum = df.groupby(['Item', 'Location'])['Price Per Unit'].sum().reset_index()
print(grouped_sum)
```

```python
grouped_sum = df.groupby(['Item' ,'Location'])['Price Per Unit'].sum().mean()
print(grouped_sum)
```

#### Aggregate multiple functions

```python
grouped_agg = df.groupby('Location')['Price Per Unit'].agg(['mean', 'sum', 'count'])
grouped_agg
```

👉 “Group data by Location, and for each location, calculate:

* average price
* total price
* number of entries”

.agg means instead of just one operation, you **apply multiple operations** at once:

| Function | Meaning                   |
| -------- | ------------------------- |
| mean     | average                   |
| sum      | total                     |
| count    | number of non-null values |


### Merging and joining Dataframes

```python
df1 = pd.DataFrame({'Key' : ['A', 'B', 'C'], 'Value1' : [1, 2, 3]})
df2 = pd.DataFrame({'Key' : ['A', 'B', 'D'], 'Value2' : [4, 5, 6]})
```

```python
df1
```

```python
df2
```

```python

pd.merge(df1, df2)
```

```python
# Merge DataFrame on the 'key columns'
pd.merge(df1, df2, on = "Key", how ="inner") 
```

👉 “Combine df1 and df2 using the common column ‘Key’, and keep only matching rows.”


* on="Key"

Think of it like:

“Join rows where df1.Key == df2.Key”

* how="inner"
 
👉 This decides what type of merge

INNER JOIN means:

Only keep rows where the key exists in BOTH dataframes

* Other how options

| Type  | Meaning            |
| ----- | ------------------ |
| inner | only matching rows |
| left  | all rows from df1  |
| right | all rows from df2  |
| outer | all rows from both |


Example intuition:

left → “I trust df1, fill from df2 if possible”

outer → “give me EVERYTHING”





```python
#If column names are different:

pd.merge(df1, df2, left_on="Value1", right_on="Value2")
```

```python
pd.merge(df1, df2, on = "Key", how ="left") 
```

```python
pd.merge(df1, df2, on = "Key", how ="right") 
```

```python
pd.merge(df1, df2, on = "Key", how ="outer") 
```
