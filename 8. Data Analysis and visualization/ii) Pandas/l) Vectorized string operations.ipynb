```python
pd.Timestamp('2026')
```

```python
import datetime as dt
```

```python
dt.datetime(2026, 5, 5, 1, 30, 26)
```

both are different objects of python libraries but serve same purpose

```python
pd.Timestamp(dt.datetime(2026, 5, 5, 1, 30, 26))
```

```python
x = pd.Timestamp(dt.datetime(2026, 5, 5, 1, 30, 26))
```

```python
# main benefite --- can fatch any attribute
print(x.year)
print(x.month)
print(x.day)
print(x.hour)
print(x.minute)
print(x.second)
```

datetime is vv convenient but performance takes a hit when performing with huge data

the weakness of python being slow programming language so they built completely new datatype (datetime64 of numpy) to hold date type values and internally those are stored as integers

it supports vectorized operation

### to_datetime function

converts an existing objects to pandas timestamp/datetimeindex object

```python
s = pd.Series(['2026/5/5', '2026/6/6', '2026/7/7'])
s.str.split('/').str.get(0)
```

```python
pd.to_datetime(s)
```

```python
pd.to_datetime(s).dt.year
```

```python
pd.to_datetime(s).dt.day_name()
```

```python
s = pd.Series(['2026/5/5', '2026/6/6', '2026/70/7'])
pd.to_datetime(s)
```

```python
pd.to_datetime(s , errors='coerce')
```

NaT = Not a Time

```python
pd.to_datetime(s , errors='coerce').dt.month_name()
```
