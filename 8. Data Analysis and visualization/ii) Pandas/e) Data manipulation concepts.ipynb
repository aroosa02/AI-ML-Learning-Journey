## Implement

```python
import pandas as pd
import numpy as np
data = {
    "name": ["Alice", "Bob", "Charlie", "Alice", "Bob", "Eve", None],
    "score": [90, 85, 90, 90, 85, 70, np.nan],
    "year": [2021, 2020, 2021, 2021, 2020, 2019, 2022],
    "city": ["NY", "LA", "NY", "NY", "LA", None, "SF"]
}

df = pd.DataFrame(data)
df
```

```python
print( df["name"].value_counts())
print()
print( df.value_counts())
```

```python
print( df.sort_values("year"))
print()
print(df.sort_values("year", ascending=False))
```

```python
df["score_rank"] = df["score"].rank(ascending=False)
df
```

```python
print(df.isnull())
print()
print(df.dropna())
print()
print(df.dropna(how="all"))
print()
df_filled = df.fillna(0)
print(df_filled)
```

dropna(how="all") means:

* It removes only those rows where all values are missing (NaN)

* If even one value exists, the row stays.

1. Remove rows with any missing values

> df.dropna()

2. Remove rows where specific column is missing

> df.dropna(subset=["name"])

3. Remove columns with all NaN

> df.dropna(axis=1, how="all")

## Column manipulation

```python
import pandas as pd
import numpy as np
data = {
    "student_id": [101,102,103,104,105,106,107,108],
    "name": ["Ali","Sara","Ahmed","Zara","Usman","Hina","Bilal","Ayesha"],
    "age": [21,22,20,23,24,21,22,20],
    "city": ["Lahore","Karachi","Islamabad","Lahore","Karachi","Lahore","Islamabad","Karachi"],
    "math_score": [78,88,92,65,70,85,90,60],
    "english_score": [80,75,85,60,72,88,91,65],
    "study_hours": [3,5,6,2,3,4,5,2]
}

df1 = pd.DataFrame(data)
df1
```

```python
df1["total_score"] = df1["english_score"] + df1["math_score"]
df1["avg_score"] = df1["total_score"] / 2
df1["is_toper"] = df1["avg_score"] > 90
df1
```

```python
# A col using Category 
# age >= 22 → "Senior"
# age < 22 → "Junior"

# use np.where(condition, if true, if false)
import numpy as np
df1["age_category"] = np.where(df1["age"] >= 22, "Senior",  "Junior")
df1
```

```python
# Droping Col
df1.drop("student_id", axis=1)
```

```python
df1.drop("name", axis=1)
```

By default pd functions takes row eg. drop('name') compiler will think of it as row so mentioning     axis = 1 is necessary, if you don't want to mention axis you can write

df1.drop(columns=["student_id"])

df1.drop(columns=["name"])

```python
# Row Filtering
# keep only students with Scores > 70

df1 = df1[df1["math_score"] > 70]
df1
```

👉 1. Inner df1

df1["math_score"] > 70

This is used to create a condition

It checks the "math_score" column of df1

Output is a Boolean mask like:

[True, False, True, ...]

👉 2. Outer df1

df1[ ... ]

This uses that Boolean mask to filter rows

It says:

“From df1, keep only rows where condition is True”

🧠 So the full meaning becomes:

“From df1, keep only those rows of df1 where math_score > 70”

🔍 Think of it like:

Inner df1 → checks

Outer df1 → filters

💥 Simple analogy

Imagine a list of students:

Inner part → “Who scored above 70?” ✅❌

Outer part → “Okay, now only show those students”
