```python
import numpy as np
a = np.arange(12)
a = a.reshape(3,4)
print(a)
a = a.flatten()
print(a)


# stacking
b = np.array([1,2,3])
c = np.array([4,5,6])

print(np.vstack((b,c)))
print(np.hstack((b,c)))
```

### random and statistics

```python
a = np.random.rand(3)
print(a)
b = np.random.randn(3)
print(b)
c = np.random.randint(1,10,5)
print(c)

# statistics
a = np.array([1,2,3,4,5])
print(np.mean(a))
print(np.std(a))
```
