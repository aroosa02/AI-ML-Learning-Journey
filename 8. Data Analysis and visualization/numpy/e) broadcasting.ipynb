# Broadcasting

Broadcasting in NumPy is a smart way of performing operations on arrays of different shapes (sizes) without needing to manually resize them.

Instead of forcing arrays to be the same size, NumPy automatically “stretches” the smaller array so it can match the bigger one during operations like addition, multiplication


Broadcasting Rules (Very Important)

NumPy compares shapes from right to left:

* If dimensions are equal → ✅ OK
* If one dimension is 1 → ✅ Stretch it
* Otherwise → ❌ Error

```python
import numpy as np

a = np.array([1, 2, 3])
b = 2

print(a + b)
```

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20, 30])

print(a + b)
```

```python


a = np.random.randint(1,20, 10) 
print(a)
print()
```

```python
# 2D + 1D
a = np.array([[1,2,3],[4,5,6]])
b = np.array([1,2,3])
print(a + b)
```

```python
a = np.array([[1,2,3],[4,5,6]])
b = np.array([10,20,30])

print(a + b)
```

```python
a = np.array([[1],[2],[3]])
b = np.array([10,20,30])

print(a + b)
```

### Broadcasting with Column

```python
a = np.array([[1],
              [2],
              [3]])

b = np.array([10, 20, 30])

print(a + b)
```

Column + row → creates a full matrix

```python
a = np.array([1, 2, 3])
b = np.array([1, 2])

print(a + b)
```

broadcasting fails because shapes don't match.

```python
a = np.arange(12).reshape(4,3)  
b = np.arange(12).reshape(3,4)  # there is no one to increment and match shape in both

a - b
```

```python
a = np.arange(3).reshape(1,3)  # ( 1, 3 ) -->> (1->2->3 , 3) -->> (3,3)
b = np.arange(3).reshape(3,1)  # ( 3, 1 ) -->> (3,  1->2->3) --->> (3,3)
#                                  increment the one and match its coresponding nmbrs 

print( a* b )
```
