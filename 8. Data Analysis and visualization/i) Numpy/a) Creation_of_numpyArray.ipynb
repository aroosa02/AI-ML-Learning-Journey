# Numpy

Numpy is a fundamental library for scientific computing in python. It provides support for arrays and matrices, along with a collection of mathematical functions to operate on these data structures.


                   Structure	Example

                      Scalar	5

                      Vector	1D [1,2,3]

                      Matrix	2D [[1,2],[3,4]]

                      Tensor	3D+ arrays


```python
import numpy as np

#Create arrays using numpy
#Create a 1D array
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)
print(type(arr1))
print(arr1.shape)
```

### Attributes of numpy arrays

```python


arr = np.array([[1, 2, 3], [4, 5, 6]])

print("Array:\n", arr) 
print("Shape:", arr.shape)  #Output: (2, 3)
print("Number of dimensions:", arr.ndim)  #Output: 2
print("Total number of elements:", arr.size)  #Output: 6
print("Data type:", arr.dtype)  #Output: int64 (or int32 depending on the system)
print("Item size (in bytes):", arr.itemsize)  #Output: 8 (or 4 depending on the system)
```
