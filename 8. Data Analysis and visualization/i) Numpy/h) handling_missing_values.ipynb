# Handling missing values

In NumPy, missing values are usually represented as:

      np.nan → Not a Number

      np.isnan -> returns True for Mising Value
Task	NumPy Function/Method
* Check for NaN             | ---  np.isnan()
* Drop NaN	                | --- data[~np.isnan(data)]
* Replace NaN with mean     | --- np.nanmean() + np.where()
* Replace NaN with median   | --- np.nanmedian() + np.where()
* Replace NaN with constant	| ---  np.nan_to_num()
* Compute mean ignoring NaN | --- np.nanmean()
* Compute sum ignoring NaN	| ---  np.nansum()

```python
import numpy as np
```

```python
a = np.array([1,2,3,np.nan,4,5,np.nan])
print(a)  # nan is missing Value

mask = np.isnan(a)
print(mask)

print()

clean_data = a[~np.isnan(a)] # Boolean Indexing 
print(clean_data)
```

```python
# Replace Missing Data with Mean
data = np.array([1, 2, np.nan, 4, 5, np.nan])
print(data)

mean_val = np.nanmean(data) # # np.nanmean ignores NaN and Computes mean
print("Mean : " ,mean_val)

# replace nan with mean_val = 3 
filled_data = np.where(np.isnan(data), mean_val, data)
print("Filtered Data : ", filled_data)
```

```python
# Replace with Median
data = np.array([1, 2, np.nan, 4, 5, np.nan])
print(data)

median_val = np.nanmedian(data)
print("Median : ", mean_val)

filled_data = np.where(np.isnan(data), median_val, data)
print("Filtered Data : ", filled_data)
```

```python
# Filld Data with Constant ( a user Given Nmbr )
data = np.array([1, 2, np.nan, 4, 5, np.nan])
print(data)

filled_data_zero = np.nan_to_num(data, nan = 0) # replace with zero
print("Data after filling with 0:", filled_data_zero)
```
