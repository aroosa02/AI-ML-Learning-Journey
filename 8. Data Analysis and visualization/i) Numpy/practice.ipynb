```python
# Create a python function named date_array that accepts two dates as string format and returns a numpy array of dates between those 2 dates. The function only accept 2 strings, 
# otherwise raise error. The date format should be like this only: 2022-12-6. 
# The end date should be included and for simplicity, choose dates from a same year.

# Example 1:

# Input:

# date_array(start = '2020-09-15', end = '2020-09-25')
# Output:

# ['2020-09-15', '2020-09-16', '2020-09-17', '2020-09-18',
#  '2020-09-19', '2020-09-20', '2020-09-21', '2020-09-22',
#  '2020-09-23', '2020-09-24', '2020-09-25']
import numpy as np

def date_array(start, end):

    # type checking
    if not isinstance(start, str) or not isinstance(end, str):
        raise TypeError("Both inputs must be strings")

    start_date = np.datetime64(start)
    end_date = np.datetime64(end)

    # include end date
    dates = np.arange(start_date, end_date + np.timedelta64(1, 'D'))

    return dates

print(date_array('2020-09-15', '2020-09-25'))
```

```python
# Fetch Every alternate column of the array
# Normalise the given array
import numpy as np

# given array
arr1 = np.random.randint(low=1, high=10000, size=40).reshape(8,5)

print("Original Array:\n", arr1)

#Fetch every alternate column
alternate_columns = arr1[:, ::2]
print("\nAlternate Columns:\n", alternate_columns)

#Normalize the array
xmin = np.min(arr1)
xmax = np.max(arr1)

normalized_array = (arr1 - xmin) / (xmax - xmin)

print("\nNormalized Array:\n", normalized_array)
```

```python
# Write a function which will accept 2 arguments.
# First: A 1D numpy array arr

# Second: An integer n {Please make sure n<=len(arr)}

# Output: The output should be the nth largest item out of the array

# Example1 : arr=(12,34,40,7,1,0) and n=3, the output should be 12
# Example2 : arr=(12,34,40,7,1,0) and n=1, the output should be

import numpy as np

def find_value(arr, n):
    if n > len(arr):
        print("n should be <=", len(arr))
        return None
    
    arr_sorted = np.sort(arr)[::-1]   # sort descending
    return arr_sorted[n-1]


a = np.array([12,34,40,7,1,0])
result = find_value(a,3)

print(result)
```

```python
#Write a program implement Boradcasting Rule to check if two array can be added or not.
# Given tuples of shapes.

# shape of a- (3,2,2)
# shape of b- (2,2)

def can_broadcast(shape_a, shape_b):

    a = list(shape_a[::-1])
    b = list(shape_b[::-1])

    for i in range(max(len(a), len(b))):
        
        dim_a = a[i] if i < len(a) else 1
        dim_b = b[i] if i < len(b) else 1

        if dim_a != dim_b and dim_a != 1 and dim_b != 1:
            return False

    return True


shape_a = (3,2,2)
shape_b = (2,2)

print("Broadcast possible:", can_broadcast(shape_a, shape_b))
```
