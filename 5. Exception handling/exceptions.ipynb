```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Invalid input. Please enter a valid number.")
except ZeroDivisionError:
    print("Cannot divide by zero. Please enter a number greater than zero.") 
except Exception as ex:    #parent class of all exceptions
    print(f"An unexpected error occurred: {ex}")       
```

```python
#try, except, else block
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Invalid input. Please enter a valid number.")
except ZeroDivisionError:
    print("Cannot divide by zero. Please enter a number greater than zero.") 
except Exception as ex: 
    print(f"An unexpected error occurred: {ex}")       
else:
    print(f"The result is: {result}")
    # else gets ignored if any exception is raised in the try block.
    #  It executes only when there are no exceptions.
```

```python
#try, except, else, finally block
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Invalid input. Please enter a valid number.")
except ZeroDivisionError:
    print("Cannot divide by zero. Please enter a number greater than zero.") 
except Exception as ex: 
    print(f"An unexpected error occurred: {ex}")
else:
    print(f"The result is: {result}")
finally:
    print("Execution of the try-except-else-finally block is complete.")
    #finally block executes regardless of whether an exception was raised or not.
    #  It is typically used for cleanup actions, such as closing files or releasing resources.
```

```python
#file handling and exception handling

try:
    file = open("file.txt", 'r')
    content = file.read()
    print(content)
except FileNotFoundError:
    print("The file does not exist.")    
finally:
    if 'file' in locals() and not file.closed:
        file.close()
        print("File has been closed.")
         
```

```python
def sum_list(lst):
    total = 0
    try:
        for item in lst:
            total += item
    except TypeError as e:
        print(f"Error: {e}")
        total = None
    finally:
        print("Execution complete.")
    return total
print(sum_list([1, 2, 3, 4, 5]))  # Output: 15
print(sum_list([1, 2, '3', 4, 5]))
```

```python
# write a function that takes a dictionary and key as input and returns value associated with the key. Handle the case when the key is not present in the dictionary.
def get_value_from_dict(dictionary, key):
    try:
        value = dictionary[key]
        return value
    except KeyError as e:
        print(f"Error: {e}")
        return None
    finally:
        print("Function execution complete.")

dictionary = {'a': 1, 'b': 2, 'c': 3}
print(get_value_from_dict(dictionary, 'b'))          
```

```python
#Nested exception handling
def nested_exception_handling(s):
    try:
        try:
            num = int(s)
        except ValueError as e:
            print(f"Conversion error: {e}")
            num = None
        finally:
            print("Conversion attempt complete.")
        if num is not None:
            try:
                result = 10 / num
            except ZeroDivisionError as e:
                print(f"Division error: {e}")
                result = None
            finally:
                print("Division attempt complete.")
            return result
    finally:
        print("Overall execution complete.")
nested_exception_handling("0")
nested_exception_handling("abc")        
```

```python
#exception handling for a list
def get_list_element(lst, index):
    try:
        element = lst[index]
    except IndexError as e:
        print(f"Error: {e}")
        element = None
    finally:
        print("Execution complete.")
    return element
my_list = [1, 2, 3, 4, 5]
print(get_list_element(my_list, 4))
```

```python
#Exception handling in network operations
import requests

def read_url(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.text
    except requests.RequestException as e:
        print(f"Network error: {e}")
        return None
    finally:
        print("Execution complete.")
print(read_url("https://www.example.com"))  
print(read_url("https://www.invalidurl.com"))      
```

```python
#Exception handling in json parsing
import json

def parse_json(json_string):
    try:
        data = json.loads(json_string)
        return data
    except json.JSONDecodeError as e:
        print(f"JSON error: {e}")
        return None
    finally:
        print("Execution complete.")
print(parse_json('{"name": "Aroosa", "age": 21}'))   
print(parse_json('{"name": "Aroosa", "age": 21'))    #missing closing brace     
```

```python
#Custom exception handling
class CustomError(Exception):
    pass
def check_for_negatives(lst):
    try:
        for num in lst:
            if num < 0:
                raise CustomError(f"Negative number found: {num}")
    except CustomError as e:
        print(f"Error: {e}")
    finally:
        print("Execution complete.")
print(check_for_negatives([1, 2, -3, 4, 5]))
print(check_for_negatives([1, 2, 3, 4, 5]))
```

```python
#Exception handling in list comprehension
def convert_with_comprehension(lst):
    try:
        integers = [int(item) for item in lst]
    except ValueError as e:
        print(f"Error: {e}")
        integers = None
    finally:
        print("Execution complete.")
    return integers
print(convert_with_comprehension(['1', '2', 'three', '4'])) 
print(convert_with_comprehension(['1', '2', '3', '4']))
```

```python
def write_strings_to_file(strings, filename):
    try:
        file = open(filename, 'w')
        for string in strings:
            file.write(string + '\n')
    except IOError as e:
        print(f"Error: {e}")
    finally:
        try:
            file.close()
        except NameError:
            pass


write_strings_to_file(['Hello', 'World'], 'output.txt')
```
