```python
class MyError(Exception):
    pass

try:
    raise MyError("Something went wrong!")
except MyError as e:
    print("Caught:", e)
```

```python
class AgeTooSmallError(Exception):
    pass

age = 15

try:
    if age < 18:
        raise AgeTooSmallError("You must be at least 18.")
    print("Access granted")
except AgeTooSmallError as e:
    print(e)
```

```python
class BalanceError(Exception):
    def __init__(self, balance, withdraw):
        self.balance = balance
        self.withdraw = withdraw
        super().__init__(f"Withdraw {withdraw} exceeds balance {balance}")

balance = 500

try:
    withdraw = 800
    if withdraw > balance:
        raise BalanceError(balance, withdraw)

except BalanceError as e:
    print("Transaction Failed:", e)
```

```python
class NegativeNumberError(Exception):
    pass

def square_root(x):
    if x < 0:
        raise NegativeNumberError("Cannot calculate square root of negative number")
    return x ** 0.5

try:
    print(square_root(-4))
except NegativeNumberError as e:
    print("Error:", e)
```

```python
class TooShortPassword(Exception):
    pass

class NoDigitPassword(Exception):
    pass

password = "abc"

try:
    if len(password) < 6:
        raise TooShortPassword("Password must be at least 6 characters")

    if not any(char.isdigit() for char in password):
        raise NoDigitPassword("Password must contain a number")

except TooShortPassword as e:
    print(e)

except NoDigitPassword as e:
    print(e)
```

```python
class UserNotFound(Exception):
    pass

class WrongPassword(Exception):
    pass

username = "admin"
password = "1234"

try:
    input_user = "admin"
    input_pass = "1111"

    if input_user != username:
        raise UserNotFound("User does not exist")

    if input_pass != password:
        raise WrongPassword("Password incorrect")

    print("Login successful")

except UserNotFound as e:
    print(e)

except WrongPassword as e:
    print(e)
```

```python
class Error(Exception):
    pass
class dobException(Error):
    pass

year = int(input("Enter the dob:"))
age = 2026 - year


try:
    if age<=30 and age>=20:
        print("The age is valid so you can apply for the exam")
    else:
        raise dobException
except dobException: #catching the exception
    print("Sorry, your age should be greater than 20 or less than 30.")
```
