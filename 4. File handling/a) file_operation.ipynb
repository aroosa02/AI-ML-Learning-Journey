```python
# Read a whole file
with open('file.txt', 'r' ) as file:
    content = file.read()
    #for line in file:
    print(content)
```

```python
# Read a file line by line
with open('file.txt', 'r' ) as file:
    for line in file:  
        # print(line)     #taking new line characters
        print(line.strip())   #removing new line characters
```

```python
# Writing a file(Overriting)
with open('file.txt', 'w') as file:
    file.write('Hello World!\n')
    file.write('Welcome to Python programming.\n')
    file.write('This is a file handling example.\n')
```

```python
# Write to a file (without overwriting  )
with open('file.txt', 'a') as file:
    file.write('This line will be appended to the file.\n')
```

```python
# Writing list of lines to a file
lines = ['First line\n', 'Second line\n', 'Third line\n']
with open('file.txt', 'a') as file:
    file.writelines(lines)
```

```python
# Binary files
# Writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file: #--> wb--> write binary
    file.write(data)
```

```python
# Reading a binary file
with open('example.bin', 'rb') as file: #--> rb--> read binary
    binary_data = file.read()
    print(binary_data)
```

```python
# Read the content from a source text file and write it to a destination text file
with open('file.txt', 'r') as source_file:
    content = source_file.read()
with open('copy_file.txt', 'w') as destination_file: 
    destination_file.write(content)   
```

```python
# Read a text file and count the number of lines, words, and characters in it
with open('file.txt', 'r') as file:
    content = file.read()
    lines = content.splitlines()
    words = content.split()
    characters = len(content)
    print(f'Lines: {len(lines)}')
    print(f'Words: {len(words)}')
    print(f'Characters: {characters}')


#                                OR

def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count
file_path = 'file.txt'
lines, words, characters = count_text_file(file_path)
print(f'Lines: {lines}, Words: {words}, Characters: {characters}')

```

```python
# Writing and then reading a file to verify the content

with open('file.txt', 'w+') as file:  #--> w+ --> write and read
    file.write('This is a test file.\nIt contains multiple lines of text overwritten.\nThis is the third line.')
    # Move the cursor back to the beginning of the file
    file.seek(0)   # prints out the content of the file after writing
    content = file.read()
    print(content)
```
