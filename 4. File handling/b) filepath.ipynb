```python
#Assignment 02 
#write a function that write a list of strings to a file with each string on a new line.
def write_strings_to_file(strings_list):
    with open("output.txt", "w") as file:
        for string in strings_list:
            file.write(string + "\n")
             #file.write("\n".join(strings_list))
# Example usage
strings = ["Hello, World!", "Welcome to Python file handling.", "This is a test file."]
write_strings_to_file(strings)
```

```python
#Assignment 06
#write a function that finds and replaces all occurences of a given word in a file with another word
def find_and_replace(file_path, target_word, replacement_word):
    with open(file_path, "r") as file:
        content = file.read()
    
    updated_content = content.replace(target_word, replacement_word)
    
    with open(file_path, "w") as file:
        file.write(updated_content)
find_and_replace("output.txt", "Python", "Java")
        
```

```python
#Assignment 07
#Write a function that reads and prints the contents of a file in reverse order (line by line).
def print_file_in_reverse(file_path):
    with open(file_path, "r") as file:
        lines = file.readlines()
    
    for line in reversed(lines):
        print(line.strip())
print_file_in_reverse("output.txt")
```

```python
#Assignment 09
#Write a function that merges the contents of two files into a new file.

def merge_files(file1_path, file2_path, output_file_path):
    with open(file1_path, "r") as file1, open(file2_path, "r") as file2:
        content1 = file1.read()
        content2 = file2.read()
    
    merged_content = content1 + "\n" + content2
    
    with open(output_file_path, "w") as output_file:
        output_file.write(merged_content)
merge_files("file.txt", "output.txt", "merged_output.txt")  
#or
#  def merge_files(file_list, output_file):
#     with open(output_file, 'w') as outfile:
#        for fname in file_list:
#               with open(fname) as infile:
#                 outfile.write(infile.read() + "\n")
# merge_files(["file.txt", "output.txt"], "merged_output.txt")                
        
```

```python
#Assignment 10
#write a function that splits large file into smaller files of a specified number of lines each.
def split_file(file_name, lines_per_file):
    with open(file_name, 'r') as file:
        lines = file.readlines()
    
    for i in range(0, len(lines), lines_per_file):
            with open(f'{file_name}_part{i//lines_per_file + 1}.txt', 'w') as part_file: #// floor division
                part_file.writelines(lines[i:i + lines_per_file]) #slicing is being used
split_file('large_file.txt', 5)            
```

```python
#Assignment 11
#Write a function that creates a log file and writes log messages with timestamps.
import datetime

def log_message(log_file, message):
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open(log_file, 'a') as file:
        file.write(f'[{timestamp}] {message}\n')   

log_message("activity.log", "This is a log message.")             
```

```python
#Assignment 15
#Write a function that attempts to read a file and handle any permission errors that may occur.
def read_protected_file(filename):
    try:
        with open(filename, 'r') as file:
            print(file.read())
    except PermissionError as e:
        print(f"Permission error: {e}")
read_protected_file("file.txt")
```
