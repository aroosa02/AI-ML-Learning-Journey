# GENERATORS
subclass of iterators

A simpler way to create iterators.
They use the yield keyword to produce a series of values lazily, which means they generate values on the fly and do not store them in memory. 

They generate values one by one using yield instead of return.

* return --- ends function, gives one value
* yield --- pauses function, gives many values

Generators save memory

eg, range(1000000000)

Python doesn't store billion numbers, it generates them when needed

(works like list comprehension)

```python
def square(n):
    for i in range(3):
        return i**2
```

```python
square(3)
```

```python
def square(n):
    for i in range(3):
        yield i**2
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
a = square(3)
a #generator object
```

```python
next(a)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen = my_generator()
gen
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```

### Practical for reading large files

Generators are particularly useful for reading large files because they allow you to process one line at a time without loading the entire file into memory.

```python
def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line
```

```python
file_path ='large_file.txt'

for line in read_large_file(file_path):
    print(line.strip())
```

```python
# Filtering Data ... Imagine A data Pipeline
# numbers → filter even → square them

def numbers():
    for i in range(10):
        yield i

# filter Even

def even(nums):
    for n in nums:
        if n%2 == 0:
            yield n

#square Them
def Sq(nums):
    for n in nums:
        yield n*n

## Implementation
data = numbers()          # each value processed one-by-one
evens = even(data)
squares = Sq(evens)

for s in squares:
    print(s)
```

## Use of Iterators and Generators Together
Scenario: Processing Huge Sensor Data for IoT

Imagine a company has IoT sensors collecting temperature and humidity in thousands of factories.

Each factory has millions of readings per day.

You need to process this data for reporting and alerts.

We’ll see how generators and iterators can work together.

```python
## Using Iterator and Generators Togather
import random

def sensor_data(num_sensors):     ##Generator function
    """Yields Reading from Sensors Continuously"""

    for sensor_id in range(1, num_sensors+1):
        # generate some random readings
        yield {
            sensor_id: sensor_id,
            "temperature": random.randint(20, 40),
            "humidity": random.randint(30, 80)
        }


## Custom Iterator
class HighTempAlert:             ##Iterator class
    def __init__(self, readings, threshold):
        self.readings = readings
        self.threshold = threshold
        self.idx = 0
        self.filtered = list(readings)

    def __iter__(self):
        return self 
#     This simply tells Python:
#     "This object itself is the iterator"

    def __next__(self):
        while self.idx < len(self.filtered):
            reading = self.filtered[self.idx]
            self.idx += 1

            if reading["temperature"] > self.threshold:
                return reading
            
        raise StopIteration



data_stream = sensor_data(5)
for reading in data_stream:
    print(reading)

# Generator produces data
data_stream = sensor_data(10)

# Iterator filters data
alerts = HighTempAlert(data_stream, threshold=30)

for alert in alerts:
    print(f"ALERT! Sensor {list(alert.keys())[0]} temp={alert['temperature']}")
```
