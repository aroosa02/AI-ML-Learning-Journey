# DECORATORS
Decorators are a powerful and flexible feature in python that allows you to modify the behavior of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code.

Example:

Add logging, authentication, timing, etc.

* function copy
* closures
* decorators


```python
# Function copy
def welcome():
    return "Welcome to the advanced python course."
welcome()
```

When you copy objects in Python, sometimes data gets shared, sometimes fully duplicated.

* Shallow Copy --- copy reference of nested objects. , --- changing original will effect copied one.
* Deep Copy --- copy everything completely., --- changing original won't effect copied one.

```python
wel = welcome #copying the function
print(wel())
del welcome
print(wel())  #the copied function is still there even after deleting the original
```

A closure remembers variables from its outer function even after the outer function finishes. --- memory of a function.

Closures are used for

* data hiding

* configuration

* decorators

```python
def outer():
    message = "Hello"

    def inner():
        print(message)

    return inner

func = outer()
func()
```

```python
# Closures --- function inside a function

def main_welcome():
    msg = "Welcome"
    def sub_welcome_method():
        print("Welcome to the advance python course.")
        print(msg) #whatevr the local variable created inside main method can also be accessed in inside method
        print("I'm learning these concepts properly.")
    return sub_welcome_method    
```

```python
main_welcome()()
```

```python
# Closures --- function inside a function

def main_welcome(func):
    
    def sub_welcome_method():
        print("Welcome to the advance python course.")
        func("Welcome everyone to this tutorial.")
        print("I'm learning these concepts properly.")
    return sub_welcome_method    
```

```python
# main_welcome(print)() OR
result = main_welcome(print)
result()
```

A decorator modifies another function without changing its code. --- adding extra behavior.

```python
def decorator_function(original_function):

    def wrapper():
        print("Before function runs")
        original_function()
        print("After function runs")

    return wrapper
```

```python
def say_hello():
    print("Hello")

say_hello = decorator_function(say_hello)

say_hello()
```

Python shortcut syntax

```python
@decorator_function
def say_hello():
    print("Hello")
say_hello()    
```

```python
## Decorator

def main_welcome(func):
    
    def sub_welcome_method():
        print("Welcome to the advance python course.")
        func()
        print("I'm learning these concepts properly.")
    return sub_welcome_method    
```

```python
def course_introduction():
    print("This is an advanced python course")

course_introduction()    
```

```python
main_welcome(course_introduction)()
```

```python
@main_welcome #decorator
def course_introduction(): #gone as parameter to that function
    print("This is an advanced python course")

    #this is equivalent to
    #course_introduction = main_welcome(course_introduction)
course_introduction()  #it is present because in jupyter if the decorator function does not call the wrapper automatically, it might not print unless the wrapper runs.   
```

```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper    
```

```python
@my_decorator
def say_hello():
    print("Hello!")
```

```python
say_hello()
```

```python
## Decorators with arguments

def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator        
```

```python
@repeat(3)
def say_hello():
    print("Hello!")
say_hello()    
```

```python
import time

def timer(func):

    def wrapper():
        start = time.time()

        func()

        end = time.time()
        print("Time:", end - start)

    return wrapper
```

```python
@timer
def slow_function():
    time.sleep(2)

slow_function()
```
