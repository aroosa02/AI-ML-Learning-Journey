# ITERATORS

Advanced Python concepts that allow for efficient looping and memory management. 

Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.

(get values one by one using iterator)

Python internally uses two methods:

* __iter__() → creates iterator object

* __next__() → returns next value

When you write:
```python
for num in numbers:
    print(num)
```
Python secretly does this:

```python
iterator = iter(numbers)

while True:
    try:
        num = next(iterator)
        print(num)
    except StopIteration:
        break
```

So for loop = iterator + next()

````python
my_list = [1, 2, 3, 4, 5, 6]
for i in my_list:
    print(i)
````

````python
# String iterator

my_string = 'Hello'
string_iterator = iter(my_string)

print(next(string_iterator)) #output: H
print(next(string_iterator)) #output: e
````

````python
# Iterate through a string (an iterable)
for char in "banana":
    print(char)
````

````python
#Creating custom iterator
class EvenNumbers:
    def __iter__(self):
        self.num = 0
        return self # returns the iterator object itself
    
    def __next__(self):
        if self.num <= 10:
            current_num = self.num
            self.num += 2
            return current_num
        else:
            raise StopIteration # signals the end of iteration

# Use the custom iterator
my_evens = EvenNumbers()
my_iter = iter(my_evens)

print(next(my_iter)) # Output: 0
print(next(my_iter)) # Output: 2
print(next(my_iter)) # Output: 4

# Or, use it in a for loop:
for number in EvenNumbers():
    print(number) 
````

````python
# Reverse Iterator 
nm = [1,23,4,5,6,78]

rv_it = reversed(nm)
print(next(rv_it))
print(next(rv_it))
````
