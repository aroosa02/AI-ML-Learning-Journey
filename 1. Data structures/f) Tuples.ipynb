#A tuple in Python is a built-in data type used to store multiple values in a single variable.
#Tuples are similar to lists, but they are immutable, meaning that once a tuple is created, its elements cannot be modified, added, or removed.
#Tuples are defined using parentheses () and can contain elements of different data types, including numbers, strings, and even other tuples.

# Ordered
# Indexed
# Immutable (cannot be changed after creation)
# Allow duplicate values

```python
#Creating a tuple
my_tuple = (1, 2, 3, "Aroosa", True)
print(my_tuple)
```

```python
#Python automatically treats comma-separated values as a tuple.
my_tuple = 1, 2, 3, "Aroosa", True
print(my_tuple)
```

```python
#Accessing elements of the tuple
print(my_tuple[0]) # 1
print(my_tuple[-3]) # 3 ----> negative indexing starts from the end of the tuple
#convert to positive index
print(my_tuple[len(my_tuple)-3]) # 3  
```

#tuples are immutable, so we cannot modify them directly. However, we can create a new tuple by concatenating existing tuples or by slicing.
#or by converting the tuple to a list, modifying the list, and then converting it back to a tuple.


```python
#Concatenating tuples
tuple1 = (1, 2, 3)  
tuple2 = ("Aroosa", True)
new_tuple = tuple1 + tuple2  # Concatenating two tuples 
print(new_tuple)
```

```python
#Slicing tuples
print(new_tuple[1:4]) #print from index 1 to 3 (4 is not included)
```

```python
#Converting tuple to list and back to tuple 
temp_list = list(new_tuple)  # Convert tuple to list
temp_list[0] = 10  # Modify the list    
new_tuple = tuple(temp_list)  # Convert back to tuple
print(new_tuple)
```

```python
#tuples have 2 built-in methods: count() and index()
print(new_tuple.count(2)) # counts the number of occurrences of the specified value (2 in this case)
print(new_tuple.index("Aroosa")) # returns the index of the first occurrence of the specified value ("Aroosa" in this case)         
```

```python
#membership operators
print(10 in new_tuple) # checks if 10 is present in the tuple  
print("Aroosa" in new_tuple) # checks if "Aroosa" is present in the tuple
print(new_tuple)
```

```python
#Tuple unpacking
a, b, c, d, e = new_tuple  # Unpacking the tuple into individual variables
print(a)  # 10
#using underscore to ignore a value during unpacking
a, _, c, _, e = new_tuple  # Unpacking the tuple while ignoring certain values
print(c)  # 3
#using * to unpack remaining values into a list
a, *rest = new_tuple  # Unpacking the first value into 'a' and the rest into 'rest'
print(rest)  # [2, 3, 'Aroosa', True]----- *collects remaining values
```

```python
#nested tuples
nested_tuple = (1, 2, (3, 4), "Aroosa", True)
print(nested_tuple)
```

```python
nested_tuple = ((1,2,3), (4,5,6), ("Aroosa", True))
print(nested_tuple[0])
print(nested_tuple[1][2])
print(nested_tuple[2][0])
```

# Use tuples when:

# ✅ Data should NOT change
# ✅ Fixed configuration values
# ✅ Returning multiple values from function
# ✅ Faster than lists
# ✅ Used as dictionary keys
