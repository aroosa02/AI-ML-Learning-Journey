```python
#short and powerful way to create a list using a loop in one line.
#generating list on the fly

lst = [i for i in range(10)]
print(lst)
```

```python
lst = [i*i for i in range(10) if i%2==0] #generating list of squares of even numbers from 0 to 9
print(lst)
```

```python
marks = [20, 30, 40, 50, 60]
new_marks = []
for x in marks:
    new_marks.append(x+5) #adding 5 to each element in the marks list and storing it in new_marks list
print(new_marks)
#but using list comprehension we can achieve the same result in one line
new_marks = [x+5 for x in marks]    
print(new_marks)  #reduced line of code and improved readability
```

```python
squares = []
for i in range(1, 6):
    squares.append(i**2)

print(squares)
#using list comprehension
squares = [i**2 for i in range(1, 6)]   
print(squares)
```

```python
#get even numbers
evens = [i for i in range(1, 11) if i % 2 == 0]
print(evens)
```

```python
#if-else
numbers = [1, 2, 3, 4, 5]
result = ["Even" if i % 2 == 0 else "Odd" for i in numbers]
print(result)
```

```python
lst1 = [1, 2, 3]
lst2 = [4, 5, 6]
pair = [(x, y) for x in lst1 for y in lst2] #generating pairs of elements from two lists
print(pair)
```

```python
words = ["hello", "world", "python", "programming"]
lengths = [len(word) for word in words] #creating a list of lengths of words in the words list
print(lengths)
lengths = {word: len(word) for word in words} #creating a dictionary with words as keys and their lengths as values
print(lengths)
```
