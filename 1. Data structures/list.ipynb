```python
# Ordered
# Indexed
# Mutable (can change)
# Allows duplicate values
# Can store different data types
marks =[3,5,6,"Aroosa", True]
print(marks)
```

```python
#Accessing elements of the list
print(marks[0]) # 3
print(marks[-3]) # 6 ----> negative indexing starts from the end of the list
#convert to positive index
print(marks[len(marks)-3]) # 6
print(marks[5-3]) # 6
```

```python
#Empty list
empty_list = []
print(empty_list)
#or
empty_list = list()
print(empty_list)
```

```python
#to check if an element is present in the list
if 5 in marks:
    print("5 is present in the list")   
else:
    print("5 is not present in the list")  

if "Aro" in marks:
    print("Aro is present in the list") 
else:    print("Aro is not present in the list")     
```

```python
#Slicing
print(marks[1:]) #print all starting from index 1
print(marks[:]) #print all elements of the list
print(marks[0:3]) #print from index 0 to 2 (3 is not included)
print(marks[1:-1]) #print from index 1 to 3 (4 is not included) or second to second last element
#Jump Indexing
print(marks[0:5:2]) #print from index 0 to 4 with a step of 2 (5 is not included)
#list[start:stop:step]
#5 represents the length of the list, and 2 represents the step size, which means it will print every second element starting from index 0 up to index 4.
print(marks[::2])  # Skip one element
```

```python
#modifying list
marks[0] = 4
print(marks)
```

```python
#adding elements to the list
marks.append(0) # adds an element at the end of the list
print(marks)
marks.insert(2, 7) # adds an element at the specified index (index 2 in this case)
print(marks)
```

```python
#extend method is used to add multiple elements to the end of the list
marks.extend([8, 9, 10]) # adds multiple elements at the end of the list
print(marks)
```

```python
#removing elements from the list
marks.remove(6) # removes the first occurrence of the specified value (6 in this case)
print(marks)
```

```python
#pop method is used to remove an element at a specific index and return it. If no index is specified, it removes and returns the last element of the list.
marks = [3, 5, 6, 8, 1]
removed_element = marks.pop(2) # removes the element at index 2 and returns it 
print(marks)
print("Removed element:", removed_element)
#marks.clear() # removes all elements from the list
#print(marks)
marks.sort() # sorts the list in ascending order
print(marks)
marks.sort(reverse=True) # sorts the list in descending order
print(marks)
marks.reverse() # reverses the order of the list
print(marks)
print(marks.count(5)) # counts the number of occurrences of the specified value (5 in this case)
```

```python
#looping through the list
for mark in marks:
    print(mark)

for i in range(len(marks)):
    print(marks[i])
```
