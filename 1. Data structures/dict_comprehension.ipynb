```python
numbers = [1, 2, 3, 4]
squares = {num: num**2 for num in numbers}
print(squares)
```

```python
numbers = {
    "a": 10,
    "b": 20,
    "c": 5
}
filtered = {k: v for k, v in numbers.items() if v > 10}
print(filtered)
```

```python
#JSON and DICTIONARY

import json

# Python dictionary
user = {
    "name": "Aroosa",
    "age": 21
}

# Convert to JSON (send to server)
json_user = json.dumps(user)

# Convert back to dictionary (receive from server)
python_user = json.loads(json_user)

print(python_user["name"])
```
