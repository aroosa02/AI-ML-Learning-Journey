```python
#A set is a built-in data type in Python used to store multiple values.
#Sets are unordered, unindexed, and do not allow duplicate values.
#Sets are defined using curly braces {} or the set() constructor.
#Sets are mutable, meaning that elements can be added or removed after creation.

#Creating a set
my_set = {1, 2, 3, "Aroosa", True}
print(my_set)
```

```python
#Creating a set using set() constructor
my_set = set([1, 2, 3, "Aroosa", True])
print(my_set)
```

```python
#Accessing elements of the set (not possible with indexing since sets are unordered)
#We can check if an element is in the set using the 'in' operator
print(1 in my_set) # checks if 1 is present in the set
print("Aroosa" in my_set) # checks if "Aroosa" is present in the set
#or
for element in my_set:
    print(element) # iterating through the set and printing each element
```

```python
#Adding elements to a set
my_set.add(4)
print(my_set)
#Adding multiple elements to a set
my_set.update([5, 6, 7])
print(my_set)
```

```python
#Removing elements from a set
my_set.remove(7) # removes element 7 from the set (raises KeyError if element doesn't exist)
print(my_set)
my_set.discard(6) # removes element 6 from the set (doesn't raise error if element doesn't exist)
print(my_set)
my_set.pop() # removes and returns an arbitrary element from the set(random dince sets are unordered)
print(my_set)
my_set.clear() # removes all elements from the set
print(my_set) # prints an empty set
```

```python
#Set operations
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1.union(set2)) # returns all elements from both sets without duplicates
print(set1 | set2) # same as union
```

```python
print(set1.intersection(set2)) # returns common elements between both sets
print(set1 & set2) # same as intersection
```

```python
print(set1.difference(set2)) # returns elements in set1 but not in set2
print(set2.difference(set1)) # returns elements in set2 but not in set1
print(set1 - set2) # same as difference
```

```python
print(set1.symmetric_difference(set2)) # returns elements in either set1 or set2 but not in both
print(set1^set2) # same as symmetric_difference
```

```python
#set methods
print(set1.isdisjoint(set2)) # checks if set1 and set2 have no common elements (returns True if they are disjoint)
print(set1.issubset(set2)) # checks if set1 is a subset of set2 (returns True if all elements of set1 are in set2)
print(set1.issuperset(set2)) # checks if set1 is a superset of set2 (returns True if all elements of set2 are in set1)  
set3=set1.copy() # creates a shallow copy of set1 and assigns it to set3
print(set3)
```

```python
#frozenset is an immutable version of a set, which means that once a frozenset is created, its elements cannot be modified, added, or removed. Frozensets are defined using the frozenset() constructor.
my_frozenset = frozenset([1, 2, 3, "Aroosa", True])
print(my_frozenset)
```

```python
empty = {}
print(type(empty))  # Dictionary
#this is an empty dictionary, not an empty set.
empty = set()
print(type(empty))  #empty set
```
