```python
#Example 1
def is_strong_password(password):
    if len(password) < 8:
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char in "!@#$%^&*()-_=+[]{}|;:'\",.<>?/" for char in password):
        return False
    return True
print(is_strong_password("Aroosa123!"))  # True
print(is_strong_password("password"))  # False

```

```python
#Example 2
def calculate_total_cost(cart):
    total_cost = 0
    for item in cart:
        total_cost += item['price'] * item['quantity']
    return total_cost

cart = [
    {'name': 'Laptop', 'price': 1000, 'quantity': 1},
    {'name': 'Phone', 'price': 500, 'quantity': 2}, 
    {'name': 'Headphones', 'price': 100, 'quantity': 3} 
]
print(calculate_total_cost(cart))  # Output: 2300
```

```python
#Example 3
def is_palindrome(s):
    s = s.replace(" ", "").lower()  # Remove spaces and convert to lowercase
    return s == s[::-1]  # Check if the string is equal to its reverse
print(is_palindrome("Aroosa"))  # False
print(is_palindrome("Able was I ere I saw Elba"))  # True
```

```python
#Example 4
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(5))  # Output: 120    
```
