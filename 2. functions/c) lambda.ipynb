```python
#Lambda function
#A lambda function is an anonymous function that can have any number of arguments but only one expression. It is defined using the lambda keyword and is often used for short, simple functions that are not reused elsewhere in the code.
#Syntax: lambda arguments: expression
def addition(a, b):
    return a + b
add = lambda a, b: a + b #a,b are input
print(add(10, 20)) #"Take a,b and return a+b"
```
