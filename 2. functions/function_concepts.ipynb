```python
#Types of arguments:
# default arguments, keyword arguments, variable-length arguments, and required arguments.
# default arguments
def average(a=9, b=1): 
    print("The average is", (a+b)/2)
average() # uses default values of a and b
average(10, 20) # overrides default values with provided arguments
average(15) # just overrides default value of a, b remains 1
average(b=5) # just overrides default value of b, a remains 9
```

```python
def name(fname, mname="Amber", lname="Rose"): # default arguments
    print("Full name:", fname, mname, lname)
name("Aroosa") # uses default values of mname and lname
```

```python
#keyword arguments
def greet(name, message):
    print(f"{message}, {name}!")
greet(name="Aroosa", message="Hello") # using keyword arguments to specify the values
greet(message="Hi", name="Aroosa") # order of keyword arguments doesn't matter
```

```python
#Required arguments
#the argument must be provided when calling the function, otherwise it will raise a TypeError.
def average(a, b, c=1): 
    print("The average is", (a+b+c)/3)
average(10, 20) # a and b are provided, c uses default value
```

```python
#Variable-length arguments
def name(*name):  #positional args ---> (*) for multiple parameters in tuple
    print(type(name)) # name is a tuple
    print("Hello", name[0], name[1], name[2])
name("Aroosa", "Amber", "Rose") # using variable-length arguments to pass multiple values as a tuple
#* is used to indicate that the function can accept a variable number of positional arguments, which will be collected into a tuple named name. In this case, the function can accept any number of names as arguments, and they will be accessible as elements of the name tuple within the function.
```

```python
def name(**name): # keyword args ---> (**) for multiple parameters in dictionary
    print(type(name)) # name is a dictionary
    print("Hello", name["fname"], name["mname"], name["lname"])
name(fname="Aroosa", mname="Amber", lname="Rose")
```

```python
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
print_details(name="Aroosa", age=20, city="Karachi")
```

```python
def print_detail(*args, **kwargs):
    for val in args:
        print(f"Value: {val}")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
print_detail( 1, 2, 3, "Hello", "World", name="Aroosa", age=20)        
```

```python
def average(*numbers):
    sum = 0
    for i in numbers:
        sum = sum + i
    print("The average is", sum/len(numbers))
average(10, 20, 30) # using variable-length arguments to pass multiple values as a tuple
```

```python
#multiple return values
def calculations(a, b):
    return a+b, a-b, a*b
result = calculations(10, 5)
print(result)
```

```python
#unpacking multiple return values
def calculations(a, b):
    return a+b, a-b, a*b
sum_result, sub_result, mul_result = calculations(10, 5)
print(sum_result)
print(sub_result)
print(mul_result)
```
