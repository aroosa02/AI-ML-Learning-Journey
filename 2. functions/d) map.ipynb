```python
#map function
#The map() function applies a given function to each item of an iterable (like a list) and returns a map object (which is an iterator). You can convert the map object to a list or other iterable type if needed.
#It applies a function to every element in a list.
#Syntax: map(function, iterable) 
def square(x):
    return x * x
numbers = [1, 2, 3, 4]
result = map(square, numbers)
print(list(result))
numbers = [1, 2, 3, 4]
result = map(lambda x: x * x, numbers)
print(list(result))
```

```python
#using lambda function with map
numbers = [1, 2, 3, 4]
result = map(lambda x: x * x, numbers)
print(list(result))
```

```python
#can we map multiple iterables
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]
added_numbers = list(map(lambda x, y: x + y, numbers1, numbers2))
print(added_numbers)
```

```python
#convert list of string into integers
str_numbers = ["1", "2", "3", "4"]
int_numbers = list(map(int, str_numbers)) #---> typecasting
print(int_numbers)
```

```python
#another example
words = ['apple', 'banana', 'cherry']
upper_words = list(map(lambda x: x.upper(), words))
#or
upper_words = list(map(str.upper, words)) 
print(upper_words)
```

```python
#example
def get_name(person):
    return person['name']
people = [
    {'name': 'Alice', 'age': 30},
    {'name': 'Bob', 'age': 25}, 
    {'name': 'Charlie', 'age': 35}          
]
names = list(map(get_name, people))   
print(names)  
```
