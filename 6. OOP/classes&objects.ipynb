```python
#Instance variable and method
class Dog:
    ## Constructor
    def __init__(self, name, breed): #self: keyword that refers to the current instance of the class, it is used to access the attributes and methods of the class
        self.name = name  # instance variable
        self.breed = breed  #variable = parameter
#instance variable --> a variable that is defined inside a method and belongs to the instance of the class, it is accessed using the self keyword
        print(self)
## create the object
dog1 = Dog("Buddy", "Golden Retriever")
print(dog1)
print(dog1.name)        
print(dog1.breed)
```

```python
dog2 = Dog("Max", "Labrador Retriever")
print(dog2) 
print(dog2.name)
print(dog2.breed)
```

```python
# def a class with instance method
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed
    def bark(self): #instance method --> a method that is defined inside a class and belongs to the instance of the class, it is accessed using the self keyword
        print(f"{self.name} says woof!")     
dog1 = Dog("Buddy", "Golden Retriever")
dog1.bark() #call the instance method bark() using the object dog1 
dog2 = Dog("Max", "Labrador Retriever")
dog2.bark() #call the instance method bark() using the object dog2       
```

```python
# Modeling a bank account 
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance
    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. New balance is {self.balance}.")
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds.")
        else:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance is {self.balance}.")
    def get_Balance(self):
        print(f"Current balance is {self.balance}.")   

#Create an object of the BankAccount class
account = BankAccount("123456789", 10000) 
print(account.balance)                
```

```python
#call instance method
account.deposit(1000)
```

```python
account.withdraw(500)
```

```python
print(account.get_Balance())
```

```python
#private attributes and methods
class BankAccount:
    def __init__(self, account_number, balance):
        self.__account_number = account_number #private attribute --> an attribute that is defined with double underscores and cannot be accessed outside the class
        self.__balance = balance #private attribute
    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited {amount}. New balance is {self.__balance}.")
    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient funds.")
        else:
            self.__balance -= amount
            print(f"Withdrew {amount}. New balance is {self.__balance}.")
    def get_Balance(self):
        print(f"Current balance is {self.__balance}.")
account = BankAccount("123456789", 10000)
#print(account.__balance) #this will raise an error because the attribute __balance is private and cannot be accessed outside the class        
print(account.get_Balance()) 
```

```python
#Class inheritance
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Employee(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id
employee = Employee("Aroosa", 21, "E12345")   
print(employee.name) #inherited attribute
print(employee.age) #inherited attribute     
print(employee.employee_id)
```

```python
def __str__(self):
        return f"Employee(Name: {self.name}, Age: {self.age}, Employee ID: {self.employee_id})"
employee = Employee("Aroosa", 21, "E12345")
employee.__str__()
```

```python

class Address:
    def __init__(self, street, city, zipcode):
        self.street = street
        self.city = city
        self.zipcode = zipcode

    def __str__(self):
        return f"{self.street}, {self.city}, {self.zipcode}"

class Person:
    def __init__(self, name, address):
        self.name = name
        self.address = address


my_address = Address("Main Street 12", "Lahore", "54000")

person1 = Person("Aroosa", my_address)

print(person1.address)
```

```python
#Create a class named counter with class variable count Each time an object is created increment the count. Add a method to get the current count> Create multiple objects and print the count.
class Counter:
    count = 0 
    def __init__(self):
        Counter.count += 1
    def get_count(self):
        return Counter.count    
counter1 = Counter()
counter2 = Counter()
print(counter1.get_count())
```

```python
import math

class MathOperations:
    @staticmethod     #static method --> a method that is defined with the @staticmethod decorator and does not require an instance of the class to be called, it can be called using the class name
    def sqrt(x):
        return math.sqrt(x)

print(MathOperations.sqrt(25))
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.__length = length
        self.__width = width

    @property   #property decorator is used to define a method as a property, it allows us to access the method as an attribute
    def length(self):
        return self.__length

    @length.setter    #setter decorator is used to define a method as a setter, it allows us to set the value of the property
    def length(self, length):
        self.__length = length

    @property
    def width(self):
        return self.__width

    @width.setter
    def width(self, width):
        self.__width = width

rect = Rectangle(10, 5)
print(rect.length, rect.width)  
rect.length = 15
rect.width = 7
print(rect.length, rect.width)       
```
