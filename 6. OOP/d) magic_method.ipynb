```python
#Magic method
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} is {self.age} years old." 


    def __repr__(self):  #custom string representation
        return f"Person(name={self.name}, age={self.age})"   

person = Person("Aroosa", 21)
print(person)       
print(repr(person)) 
```

```python
class Point:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        return Point(self.x - other.x, self.y - other.y)
    
    def __mul__(self, other):
        return Point(self.x * other.x, self.y * other.y)
    
    def __eq__(self, other):
        return Point(self.x == other.x and self.y == other.y)
    
    def __repr__(self):
        return f"Points({self.x}, {self.y})"

#Create objects of the class

p1 = Point(2,3)
p2 = Point(4,5)

p3 = p1 + p2


print(p3.x, p3.y)
print(p1+p2)
print(p1*p2)
```

Here:

p1 + p2

actually calls

p1.__add__(p2)

```python
class Number:

    def __init__(self, value):
        self.value = value

    def __sub__(self, other):
        return Number(self.value - other.value)


n1 = Number(10)
n2 = Number(3)

result = n1 - n2

print(result.value)
```

Operator overloading in Python is done using magic methods (also called dunder methods). These methods define how operators behave when used with objects.

```python
# OPERATOR POLYMORPHISM   ----- DUNDER FUNCTION
class Complex:
    def __init__(self, real, imaginary):
        self.real = real
        self.imaginary = imaginary

    def showNumber(self):
        print(self.real, "i +", self.imaginary,"j" )



#Dunder functions ---> double underscore__    we define our own meaning for an operator
# normally a+b   internally in class a.__add__(b)

    def add(self, num):
        newReal = self.real + num.real
        newImg = self.imaginary + num.imaginary
        return Complex(newReal, newImg)

num1 = Complex(2, 4)
num2 = Complex(5, 9) 

num1.showNumber()
num2.showNumber()

num3 = num1.add(num2)
num3.showNumber()
```

```python
class Complex:
    def __init__(self, real, imaginary):
        self.real = real
        self.imaginary = imaginary

    def showNumber(self):
        print(self.real, "i +", self.imaginary,"j" )



#Dunder functions ---> double underscore__    we define our own meaning for an operator
# normally a+b   internally in class a __add__(b)

    def __add__(self, num):   #using dunder function
        newReal = self.real + num.real
        newImg = self.imaginary + num.imaginary
        return Complex(newReal, newImg)
    

    def __sub__(self, num):   #using dunder function
        newReal = self.real - num.real
        newImg = self.imaginary - num.imaginary
        return Complex(newReal, newImg)
    
    def __mul__(self, other):
        real_part = self.real * other.real - self.imaginary * other.imaginary
        imag_part = self.real * other.imaginary + self.imaginary * other.real 
        return Complex(real_part, imag_part)
    
    def __eq__(self, other):
        return self.real == other.real and self.imaginary == other.imaginary
    
    def __repr__(self):
        return f"{self.real}i + {self.imaginary}j"


num1 = Complex(2, 4)
num2 = Complex(5, 9) 

num1.showNumber()
num2.showNumber()

num3 = num1 + num2
num3.showNumber()
num3 = num1 - num2
num3.showNumber()

print(num1 + num2)
print(repr(num1 - num2))
print(num1 == num2)
```

```python
class Book:
    def __init__(self, pages):
        self.pages = pages

    def __add__(self, other):
        return self.pages + other.pages

b1 = Book(100)
b2 = Book(200)

print(b1 + b2)

#Normally + adds numbers.
#But here we changed its behavior for Book objects.
```
