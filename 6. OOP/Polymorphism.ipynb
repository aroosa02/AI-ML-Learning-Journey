```python
#Polymorphism is the ability of an object to take on many forms. In Python, polymorphism allows us to define methods in the child class with the same name as the methods in the parent class. The child class can provide its own implementation of the method, and when we call the method on an object of the child class, it will use the implementation defined in the child class.
#method overriding and interfaces
#A way to perform a single action in different ways

#Base class
class Animal:
    def speak(self):
        return "Sound  of animal."

#derived class
class Dog(Animal):
    def speak(self):   #overriding the speak method of the base class
        return "Woof!"
    
class Cat(Animal):
    def speak(self):
        return "Meow!" 

#Function that demonstrates polymorphism
def animal_speaks(animal):
    print(animal.speak())      
    
dog = Dog()
cat = Cat()
print(dog.speak())  
print(cat.speak()) 
animal_speaks(dog)
animal_speaks(cat) 
```

```python
#advanced method overriding (using super)
class Person:
    def intro(self):
        print("I am a person")

class Student(Person):
    def intro(self):
        super().intro()
        print("I am also a student")

s = Student()
s.intro()
```

```python
#Polymorphism with functions and methods
#Base class

class Shape:
    def area(self):
        return "The area of the figure."
    
#Derived class
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width= width
        self.height = height

    def area(self):
        return self.width *  self.height

class Circle(Shape):
    def __init__(self, radius):
            self.radius = radius

    def area(self):
         return 3.14*self.radius *self.radius

#Function that demonstrates polymorphism

def print_area(shape):
     print(f"The area is {shape.area()}")           

rectangle = Rectangle(4, 5)
circle = Circle(3)

print_area(rectangle)
print_area(circle)
```

```python
# operator polymorphism
class Complex:
    def __init__(self, real, imaginary):
        self.real = real
        self.imaginary = imaginary

    def showNumber(self):
        print(self.real, "i +", self.imaginary,"j" )



#Dunder functions ---> double underscore__    we define our own meaning for an operator
# normally a+b   internally in class a __add__(b)

    def add(self, num):
        newReal = self.real + num.real
        newImg = self.imaginary + num.imaginary
        return Complex(newReal, newImg)

num1 = Complex(2, 4)
num2 = Complex(5, 9) 

num1.showNumber()
num2.showNumber()

num3 = num1.add(num2)
num3.showNumber()
```

```python
class Complex:
    def __init__(self, real, imaginary):
        self.real = real
        self.imaginary = imaginary

    def showNumber(self):
        print(self.real, "i +", self.imaginary,"j" )



#Dunder functions ---> double underscore__    we define our own meaning for an operator
# normally a+b   internally in class a __add__(b)

    def __add__(self, num):   #using dunder function
        newReal = self.real + num.real
        newImg = self.imaginary + num.imaginary
        return Complex(newReal, newImg)
    

    def __sub__(self, num):   #using dunder function
        newReal = self.real - num.real
        newImg = self.imaginary - num.imaginary
        return Complex(newReal, newImg)

num1 = Complex(2, 4)
num2 = Complex(5, 9) 

num1.showNumber()
num2.showNumber()

num3 = num1 + num2
num3.showNumber()
num3 = num1 - num2
num3.showNumber()
```

```python
class Book:
    def __init__(self, pages):
        self.pages = pages

    def __add__(self, other):
        return self.pages + other.pages

b1 = Book(100)
b2 = Book(200)

print(b1 + b2)

#Normally + adds numbers.
#But here we changed its behavior for Book objects.
```

```python
class Payment:
    def pay(self, amount):
        pass

class CreditCard(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using Credit Card")

class PayPal(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using PayPal")

class Cash(Payment):
    def pay(self, amount):
        print(f"Paid {amount} in Cash")

def process_payment(method, amount):
    method.pay(amount)

cc = CreditCard()
pp = PayPal()
cash = Cash()

process_payment(cc, 500)
process_payment(pp, 1000)
process_payment(cash, 200)
```

```python
#Polymorphism with abstract base class --- interfaces
#Abstract Base Class---abc

from abc import ABC, abstractmethod


#Define an abstract class

class Vehicle(ABC):
    @abstractmethod   #--> like an empty method initially
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        return "Car engine started."

class MotorCycle(ABC):
    def start_engine(self):
        return "MotorCycle engine started."
    

#Function that demonstrates polymorphism
def start_vehicle(vehicle):
    print(vehicle.start_engine())    
    
#Creating objects
car = Car()
bike = MotorCycle()  

start_vehicle(car)
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        return 3.14 * self.r * self.r

class Rectangle(Shape):
    def __init__(self, l, w):
        self.l = l
        self.w = w

    def area(self):
        return self.l * self.w

shapes = [Circle(5), Rectangle(4,6)]

for shape in shapes:
    print(shape.area())
```

```python
#Method Overriding
#when a child class changes the implementation of a method that already exists in the parent class.
#Parent gives method---Child redefines it.

#Abstract Base Class
#A way to force child classes to implement specific methods.
#you cannot create object of abstract class


#Hence ABC are for definig blueprint, enforcing structure

#Normal Parent Class:

#        “Here’s default behavior. You may change it.”

#Abstract Parent Class:

#        “Here’s required behavior. You must define it.”
```
