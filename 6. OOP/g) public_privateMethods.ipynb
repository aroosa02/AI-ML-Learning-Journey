```python
#PUBLIC ---- accissible from anywhere(inside or outside the class)
class Student:
    def __init__(self, name):
        self.name = name   # public variable

    def show(self):       # public method
        print("Student:", self.name)

s1 = Student("Aroosa")

print(s1.name)
s1.show()
```

```python
#PROTECTED---- (_) but python doesn't enforce it, it just says not to use unless necessary.
class Student:
    def __init__(self):
        self._marks = 90   # protected variable

s = Student()

print(s._marks)   # still accessible
```

protected can be accessed but is discouraged

```python
#PRIVATE---(__) 
class Student:
    def __init__(self):
        self.__age = 20

s = Student()

print(s.__age)  #cannot be accessed due to variable being private.
```

# NAME MANGLING:

python doesn't truly hide variable instead it renames it internally.
 __variable becomes _ClassName__variable
 That's what name mangling is.


 It prevents accidental override in subclasses.
 It avoid name conflicts.



  class Parent:
   def __hello(self):
       print("Parent")

 class Child(Parent):
   def __hello(self):
       print("Child")

These methods won't clash because they become:

_Parent__hello
_Child__hello

```python
class Student:
    def __init__(self):
        self.__age = 20

s = Student()

print(s._Student__age)

#This prevents accidental access not intentional access.
```

```python
#PRIVATE METHODS---same rules
class BankAccount:

    def __init__(self):
        self.balance = 1000

    def __show_balance(self):
        print("Balance:", self.balance)

account = BankAccount()

account.__show_balance()
```

```python
#but with name mangling
account._BankAccount__show_balance()
```

```python
class Student:

    def __hello(self):
        print("Hello from private method")

s = Student()

s._Student__hello()
```

```python
class Student:

    def __hello(self):
        print("Private hello")

    def greet(self):
        self.__hello()

s = Student()
s.greet()
```

Public method → does main work
Private method → internal helper

```python
class BankAccount:

    def __calculate_interest(self):
        print("Interest calculated")

    def update_account(self):
        self.__calculate_interest()
        print("Account updated")
acc = BankAccount()
acc.update_account()
```

Two ways private methods can still be accessed

1️⃣ Inside the class

self.__method()

2️⃣ Outside using name mangling

object._ClassName__method()

```python
class BankAccount:

    def __init__(self, balance):
        self.__balance = balance   # private variable

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance


acc = BankAccount(1000)

acc.deposit(500)

print(acc.get_balance())
```

Here we protect balance from direct modification.

Bad practice:

acc.__balance = 100000

Good practice:

acc.deposit()
