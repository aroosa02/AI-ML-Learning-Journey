# POLYMORPHISM --- one interface, many forms

Polymorphism is the ability of an object to take on many forms. In Python, polymorphism allows us to define methods in the child class with the same name as the methods in the parent class. The child class can provide its own implementation of the method, and when we call the method on an object of the child class, it will use the implementation defined in the child class.

* method overriding
* interfaces
A way to perform a single action in different ways


Types:
* Method Overriding
* Abstraction
* Method Overloading (not direct but using dunder function)

```python

#Base class
class Animal:
    def speak(self):
        return "Sound  of animal."

#derived class
class Dog(Animal):
    def speak(self):   #overriding the speak method of the base class
        return "Woof!"
    
class Cat(Animal):
    def speak(self):
        return "Meow!" 

#Function that demonstrates polymorphism
def animal_speaks(animal):
    print(animal.speak())      
    
dog = Dog()
cat = Cat()
print(dog.speak())  
print(cat.speak()) 
animal_speaks(dog)
animal_speaks(cat) 
```

```python
#advanced method overriding (using super)
class Person:
    def intro(self):
        print("I am a person")

class Student(Person):
    def intro(self):
        super().intro()
        print("I am also a student")

s = Student()
s.intro()
```

```python
#Polymorphism with functions and methods
#Base class

class Shape:
    def area(self):
        return "The area of the figure."
    
#Derived class
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width= width
        self.height = height

    def area(self):
        return self.width *  self.height

class Circle(Shape):
    def __init__(self, radius):
            self.radius = radius

    def area(self):
         return 3.14*self.radius *self.radius

#Function that demonstrates polymorphism

def print_area(shape):
     print(f"The area is {shape.area()}")           

rectangle = Rectangle(4, 5)
circle = Circle(3)

print_area(rectangle)
print_area(circle)
```

```python
class Payment:
    def pay(self, amount):
        pass

class CreditCard(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using Credit Card")

class PayPal(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using PayPal")

class Cash(Payment):
    def pay(self, amount):
        print(f"Paid {amount} in Cash")

def process_payment(method, amount):
    method.pay(amount)

cc = CreditCard()
pp = PayPal()
cash = Cash()

process_payment(cc, 500)
process_payment(pp, 1000)
process_payment(cash, 200)
```

```python
#Polymorphism with abstract base class --- interfaces
#Abstract Base Class---abc

from abc import ABC, abstractmethod


#Define an abstract class

class Vehicle(ABC):
    @abstractmethod   #--> like an empty method initially
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        return "Car engine started."

class MotorCycle(ABC):
    def start_engine(self):
        return "MotorCycle engine started."
    

#Function that demonstrates polymorphism
def start_vehicle(vehicle):
    print(vehicle.start_engine())    
    
#Creating objects
car = Car()
bike = MotorCycle()  

start_vehicle(car)
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        return 3.14 * self.r * self.r

class Rectangle(Shape):
    def __init__(self, l, w):
        self.l = l
        self.w = w

    def area(self):
        return self.l * self.w

shapes = [Circle(5), Rectangle(4,6)]

for shape in shapes:
    print(shape.area())
```

# METHOD OVERLOADING


In Java, you can do this:

add(int a, int b)
add(int a, int b, int c)

But in Python this is NOT supported directly.
Python will only keep the last method

```python
# METHOD OVERLOADING
class Calc:

    def add(self, a, b):
        return a + b
    
    # python doesnt allow this type of overLoading
    def add(self, a, b, c):
        return a + b + c


c1 = Calc()
c1.add(2,3)
c1.add(2,3,5)
```

```python
class Math:

    def add(self, a, b, c=0):
        print(a + b + c)

m = Math()

m.add(2,3)
m.add(2,3,4)

#Using default arguments polymorphism works
```

```python
class Math:

    def add(self, *nums): #variable args
        print(sum(nums))

m = Math()

m.add(2,3)
m.add(2,3,4)
m.add(1,2,3,4,5) 
```

```python
class Calculate:

    # use defalt values
    def add(self, a,b = 0,c = 0):
        if b == 0 and c == 0:
            return a
        elif b == 0:
            return a + c
        elif c == 0:
            return a + b
        else:
            return a+b+c
        
        # now this handles all the cases 
            

c1 = Calculate()
print(c1.add(1,2))
print(c1.add(4,0,7))
print(c1.add(37,56,232))

# OverLoading Works 
        
```
