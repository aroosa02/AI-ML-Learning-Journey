# AGGREGATION

Aggregation is a "has-a" relationship between two classes where:

* One class contains an object of another class

* But both objects can exist independently

So if the container object is destroyed, the other object still exists.

This is different from composition, where the contained object cannot exist without the parent.

###### Composition --- strong relation --- cannot exist independently

###### Aggregation --- weak relation --- can exist independently

```python
class Professor:
    
    def __init__(self, name):
        self.name = name


class Department:
    
    def __init__(self, prof):
        self.professor = prof   # Aggregation


p1 = Professor("Dr. Ahmed")

dept = Department(p1)

print("Professor in department:", dept.professor.name)
```

Department uses the Professor object, but doesn't create it.
This is the key idea of aggregation.

```python
#Composition
class Room:
    
    def __init__(self, name):
        self.name = name


class House:
    
    def __init__(self):
        self.room1 = Room("Bedroom")
        self.room2 = Room("Kitchen")


h = House()

print("Room 1:", h.room1.name)
print("Room 2:", h.room2.name)
```

Now we'll see how it is different from aggregation

```python
#Aggregation
class Engine:
    def start(self):
        print("Engine started")


class Car:
    def __init__(self, engine):
        self.engine = engine   # object passed from outside


e = Engine()      # created outside   
c = Car(e)        # passed into Car   <===

c.engine.start()
```

The Engine object already existed and was given to Car.

✔ Engine can exist without Car

✔ So this is Aggregation

```python
#Composition
class Engine:
    def start(self):
        print("Engine started")


class Car:
    def __init__(self):
        self.engine = Engine()   # created inside  <===


c = Car()
c.engine.start()
```

The Car creates the Engine itself.

✔ Engine depends on Car

✔ So this is Composition

### Aggregation

engine = Engine()

car = Car(engine)

Object created outside

Engine  → given to → Car


### Composition

self.engine = Engine()

Object created inside

Car  → creates → Engine

```python
#Another example code of aggregation

# Student Has Laptop
class Student:

    def __init__(self, name, roll, laptop):
        self.name = name
        self.roll = roll
        self.laptop = laptop   # Laptop object passed


    # print laptop info
    def print_laptop(self):
        print(self.laptop.brand, self.laptop.get_processor(), self.laptop.ram)


    # edit laptop info
    def upgrade_laptop(self, new_brand, new_processor, new_ram):
        self.laptop.edit_laptop(new_brand, new_processor, new_ram)



# second class
class Laptop:

    def __init__(self, brand, processor, ram):
        self.brand = brand
        self.__processor = processor   # private attribute
        self.ram = ram


    # getter for processor
    def get_processor(self):
        return self.__processor


    # edit laptop
    def edit_laptop(self, new_brand, new_processor, new_ram):
        self.brand = new_brand
        self.__processor = new_processor
        self.ram = new_ram



# Laptop object created independently
l1 = Laptop("Dell", "i5", "8GB")


# Student object
s1 = Student("Aroosa", 101, l1)

# print laptop
s1.print_laptop()


# upgrade laptop
s1.upgrade_laptop("HP", "i7", "16GB")

s1.print_laptop()
```

```python
class Pen:
    def write(self):
        print("Writing")

class Student:
    def __init__(self, pen):
        self.pen = pen


p = Pen()
s = Student(p)

s.pen.write()
```

Here:

s.pen → Pen object

So Python executes:

Pen.write()
