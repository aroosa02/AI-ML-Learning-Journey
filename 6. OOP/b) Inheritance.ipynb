```python
#INHERITANCE

#Parent class
class Car:
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def drive(self):
        print(f"The person will drive {self.enginetype} car.")    
```

```python
car1 = Car(4, 4, "Petrol")
car1.drive()
```

```python
class Tesla(Car):
    def __init__(self, windows, doors, enginetype, autopilot):
        super().__init__(windows, doors, enginetype) #super() is used to call the parent class constructor and initialize the inherited attributes
        self.autopilot = autopilot
    def autopilot_feature(self):
        if self.autopilot:
            print("The car has autopilot feature.")
        else:
            print("The car does not have autopilot feature.")
                
```

```python
tesla1 = Tesla(4, 4, "Electric", True)
tesla1.drive()
```

```python
tesla1.autopilot_feature()
```

```python
#Multiple Inheritance
#When a class inherits from more than one parent class, it is called multiple inheritance. The child class can access the attributes and methods of all the parent classes.

class Animal:   #Base class 1
    def __init__(self, name):
        self.name = name
    def speak(self):
        print("Subclass must implement this method.")   
class Pet:      #Base class 2
    def __init__(self, owner):
        self.owner = owner
    def play(self):
        print("The pet is playing.")
class Dog(Animal, Pet):   #Derived class
    def __init__(self, name, owner):
        Animal.__init__(self, name)  #Initializing the attributes of the parent classes
        Pet.__init__(self, owner)
        #or
        #super().__init__(name)  #Using super() to initialize the attributes of the parent class, it will call the __init__ method of the first parent class in the order of inheritance
    def speak(self):
        print(f"{self.name} says meow!")

cat = Dog("Kristy", "Aroosa")
cat.speak()    
print(f"Owner: {cat.owner}")      
```
