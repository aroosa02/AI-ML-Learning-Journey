```python
#Data serialization
import json
data = {'name': 'Aroosa', 'age': 21, 'city': 'Islamabad'}

json_string = json.dumps(data) # Serialize to JSON string
print(json_string)
print(type(json_string))

parsed_data = json.loads(json_string) # Deserialize back to dictionary
print(parsed_data)
print(type(parsed_data))
```

```python
# csv
import csv
with open('example.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Name', 'Age', 'City'])
    writer.writerow(['Aroosa', 21, 'Islamabad'])

with open('example.csv', mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)  
```

```python
# datetime
from datetime import datetime, timedelta
now = datetime.now()
print("Current date and time:", now)

yesterday = now - timedelta(days=1)
print("Yesterday's date and time:", yesterday)
```

```python
from datetime import datetime, timedelta

# Current date
today = datetime.now()
print("Current Date:", today.date())

# Date after 100 days
future_date = today + timedelta(days=100)
print("Date after 100 days:", future_date.date())

# Day of the week
print("Today is:", today.strftime("%A"))
```

```python
#time
import time
print("Current time:", time.time()) #ctime in seconds since the epoch
time.sleep(2) # Sleep for 2 seconds
print("2 seconds later:", time.time())
```

```python
#Regular expressions ---> search, match and find patterns inside strings (search tool)
import re
pattern = r'\b\w+ain\b' # Words ending with 'ain'  (r--> raw string telling not to treat \ as escape character)
#\b --> word boundary, \w+ --> one or more word characters, ain --> literal string 'ain', \b --> word boundary
text = "The rain in Spain stays mainly in the plain."
matches = re.search(pattern, text)
#matches = re.findall(pattern, text) # Find all matches in the text
if matches:
    print("Match found:", matches.group())
```
