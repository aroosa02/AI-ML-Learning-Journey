## Control Flow
conditionals and loops

* conditional statements
```python
if condition:
  statement
elif condition:
  statement
else:
  statement
  ```

````python
# Conditionals

age = 18

if age>= 18:
    print("You can vote!")

elif age>=16:
    print("You can drive.")

else:
    print("Too young!")        
````

````python
#Loops
for i in range(11):   #(start, stop)
    print(i)
````

In for loop, 
 
* one argument --- (stop)
* two argument --- (start, stop)
* three argument --- (start, stop, step/jump)

````python
# for loop
for i in range(3, 31, 3): #table of 3  --- (start, stop, step)
    print(i)
````

````python
for i in range(2, 8, 2):
    print(i)
````

````python
for char in "Python":
    print(char)
````

while condition:

  statement/Code to execute
  
  update condition variable inc/dec to avoid infinite loop

````python
# while loop---Unknown no. of iterations
i = 3
while i <= 30:
    print(i)
    i=i+3
````

````python
# Count down from 5

count = 5


while count > 0:

    print(count)

    count -= 1


print("Blast off!")
````

````python
count = 1
while count <= 4:
    print(count)
    count += 1
````

## Data Structures
list, tuple, set, dictionary

````python
#list --- ordered, changeable, allows duplicates

fruits = ["apple", "banana"]

fruits.append("orange")

print(fruits[0])
````

````python
data = [10, 20, 30, 40, 50]
print(data[1:4])
````

````python
#Tuples --- ordered, unchangeable, allows duplicates
point = (10, 20)

x, y = point

# Cannot modify:

# point[0] = 15  # Error!
````

````python
student = ("Ali", 21, "CS", 3.7)
print(student[1])
student[1] = 22  #it will display error as tuple is immutable
````

````python
# Sets --- unordered, no duplicates
numbers = {1, 2, 3, 2}

print(numbers)  # {1, 2, 3}

numbers.add(4)
print(numbers)
````

````python
#Dictionary --- key-value pairs, labelled dataset
student = {

    "name": "Aroosa",

    "age": 21

}

print(student["name"])
````

````python
student = {
    "name" : "Sara",
    "age" : 20,
    "age" : 21
}
print(student["age"])
print(student.get("phone"))

# Checking all keys and values
print(student.keys())         # all keys printed only
print(student.values())       # all values printed only
print(student.items())        # key-value pairs printed
````
